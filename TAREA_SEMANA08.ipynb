```python
# ============================================================
# PARTE 2 : DERIVACIÓN NUMÉRICA
# DIFERENCIAS CENTRADAS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# ============================================================
# CONTENEDOR GENERAL
# ============================================================

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 11

# ============================================================
# TÍTULO
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:Helvetica;
font-size:12pt;
color:white;
background-color:rgb(15,150,103);
padding:8px 14px;
font-weight:bold;
display:inline-block;
border-bottom:4px solid rgb(15,150,103);
margin-bottom:15px;
">

PARTE 2: DERIVACIÓN NUMÉRICA POR DIFERENCIAS CENTRADAS

</div>

</div>

"""))

#
# ============================================================
# SOLUCIÓN
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:'Dutch 801';
font-size:12pt;
font-style:italic;
font-weight:bold;
color:rgb(15,150,103);
background-color:rgb(247,239,219);
padding:8px 14px;
display:block;
border-bottom:3px solid rgb(15,150,103);
margin-bottom:20px;
">

SOLUCIÓN

</div>

</div>

"""))
# ============================================================
# DATOS
# ============================================================

t = np.array([
0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,
5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,
10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,
14.0,14.5,15.0
])

V = np.array([
4.8120,4.2635,3.7541,3.2804,2.8387,2.4236,
2.0314,1.6583,1.3011,1.2574,1.1460,0.8624,
0.6048,0.3720,0.1672,-0.0186,-0.1805,-0.3142,
-0.3148,-0.2769,-0.2100,0.0483,0.2931,0.5185,
0.7214,0.8982,1.0456,1.1618,1.2465,1.3001,1.3228
])

# ============================================================
# DIFERENCIAS CENTRADAS
# ============================================================

h = 0.5

def derivada_centrada(indice):
    return (V[indice+1] - V[indice-1])/(2*h)

# ============================================================
# ÍNDICES
# ============================================================

i4 = np.where(t == 4.0)[0][0]
i8 = np.where(t == 8.0)[0][0]
i12 = np.where(t == 12.0)[0][0]

# ============================================================
# DERIVADAS
# ============================================================

d4 = derivada_centrada(i4)
d8 = derivada_centrada(i8)
d12 = derivada_centrada(i12)

# ============================================================
# DESARROLLO TEÓRICO
# ============================================================

print("La tasa de cambio del voltaje del biosensor")
print("permite evaluar la respuesta dinámica del sistema.")
print("Para ello se emplea el método de diferencias centradas.\n")

print("FÓRMULA UTILIZADA:\n")

print("V'(t_i) = [V(t_i+h) - V(t_i-h)] / (2h)\n")

print(f"Tamaño del paso: h = {h} ms")
print(f"Entonces: 2h = {2*h} ms")

# ============================================================
# CÁLCULO 1
# ============================================================

print("\n" + "="*70)
print("1. CÁLCULO EN t = 4.0 ms")
print("="*70)

print("\nDatos vecinos:")

print(f"V(3.5 ms) = {V[i4-1]:.4f} V")
print(f"V(4.5 ms) = {V[i4+1]:.4f} V")

print("\nAplicando diferencias centradas:\n")

print("V'(4.0) = [V(4.5)-V(3.5)] / 1.0\n")

print(f"V'(4.0) = ({V[i4+1]:.4f} - {V[i4-1]:.4f}) / 1.0\n")

print(f"V'(4.0) = {d4:.4f} V/ms")

print("\nINTERPRETACIÓN FÍSICA:")

print("La derivada es negativa.")
print("El voltaje disminuye en esta región.")
print("Esto representa una relajación mecánica")
print("o disminución de la vibración detectada.")

# ============================================================
# CÁLCULO 2
# ============================================================

print("\n" + "="*70)
print("2. CÁLCULO EN t = 8.0 ms")
print("="*70)

print("\nDatos vecinos:")

print(f"V(7.5 ms) = {V[i8-1]:.4f} V")
print(f"V(8.5 ms) = {V[i8+1]:.4f} V")

print("\nAplicando diferencias centradas:\n")

print("V'(8.0) = [V(8.5)-V(7.5)] / 1.0\n")

print(f"V'(8.0) = ({V[i8+1]:.4f} - {V[i8-1]:.4f}) / 1.0\n")

print(f"V'(8.0) = {d8:.4f} V/ms")

print("\nINTERPRETACIÓN FÍSICA:")

print("La derivada es negativa.")
print("La señal continúa decreciendo")
print("cerca del valor mínimo del voltaje.")
print("Esto indica una posible transición")
print("del circuito biomédico.")

# ============================================================
# CÁLCULO 3
# ============================================================

print("\n" + "="*70)
print("3. CÁLCULO EN t = 12.0 ms")
print("="*70)

print("\nDatos vecinos:")

print(f"V(11.5 ms) = {V[i12-1]:.4f} V")
print(f"V(12.5 ms) = {V[i12+1]:.4f} V")

print("\nAplicando diferencias centradas:\n")

print("V'(12.0) = [V(12.5)-V(11.5)] / 1.0\n")

print(f"V'(12.0) = ({V[i12+1]:.4f} - {V[i12-1]:.4f}) / 1.0\n")

print(f"V'(12.0) = {d12:.4f} V/ms")

print("\nINTERPRETACIÓN FÍSICA:")

print("La derivada es positiva.")
print("El voltaje aumenta en esta región.")
print("Esto representa recuperación")
print("de la señal fisiológica.")
print("El tejido vuelve a deformarse")
print("generando mayor energía eléctrica.")

# ============================================================
# TABLA RESUMEN
# ============================================================

tabla = pd.DataFrame({

    "Tiempo (ms)":[4.0,8.0,12.0],

    "Derivada V'(t) [V/ms]":[
        round(float(d4),4),
        round(float(d8),4),
        round(float(d12),4)
    ],

    "Comportamiento":[
        "Decreciente",
        "Decreciente",
        "Creciente"
    ]
})

print("\n" + "="*70)
print("TABLA RESUMEN")
print("="*70 + "\n")

print(tabla)

# ============================================================
# GRÁFICA
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(
    t,
    V,
    marker='o',
    linewidth=2,
    label='Voltaje V(t)'
)

# Punto 1
plt.scatter(
    4.0,
    V[i4],
    s=150
)

plt.annotate(
    "1",
    (4.0, V[i4]),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

# Punto 2
plt.scatter(
    8.0,
    V[i8],
    s=150
)

plt.annotate(
    "2",
    (8.0, V[i8]),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

# Punto 3
plt.scatter(
    12.0,
    V[i12],
    s=150
)

plt.annotate(
    "3",
    (12.0, V[i12]),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

plt.axhline(
    0,
    linestyle='--'
)

plt.xlabel("Tiempo (ms)")
plt.ylabel("Voltaje V(t)")

plt.title(
    "Derivación Numérica por Diferencias Centradas",
    fontsize=14,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

# ============================================================
# GUARDAR IMAGEN
# ============================================================

nombre_imagen = "derivacion_numerica.png"

plt.savefig(
    nombre_imagen,
    dpi=300,
    bbox_inches='tight'
)

plt.show()

print("\nImagen guardada correctamente como:")
print(nombre_imagen)

# ============================================================
# MOSTRAR IMAGEN EN COLAB
# ============================================================

from IPython.display import Image, display

display(Image(filename=nombre_imagen))

# ============================================================
# CONCLUSIÓN
# ============================================================

print("\n" + "="*70)
print("CONCLUSIÓN")
print("="*70 + "\n")

print("El método de diferencias centradas permitió")
print("aproximar correctamente la derivada primera")
print("de la señal de voltaje del biosensor.")

print("\nLos resultados obtenidos muestran que:")

print("\n• En t = 4.0 ms la señal es decreciente,")
print("  indicando una disminución de la vibración.")

print("\n• En t = 8.0 ms el sistema continúa")
print("  acercándose a una zona crítica cercana")
print("  al mínimo de la señal.")

print("\n• En t = 12.0 ms la derivada positiva")
print("  representa recuperación del potencial")
print("  eléctrico detectado por el sensor.")

print("\nEl análisis numérico permite estudiar")
print("la dinámica del biosensor y evaluar")
print("el comportamiento fisiológico asociado")
print("a las microvibraciones del tejido.")

# ============================================================
# CIERRE HTML
# ============================================================

display(Markdown("""

</div>
</div>

"""))
```

```python
# ============================================================
# PARTE 3 : BÚSQUEDA DE RAÍCES POR BISECCIÓN
# SISTEMA DE TELEMETRÍA BIOMÉDICA
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# ============================================================
# CONTENEDOR GENERAL
# ============================================================
from scipy.interpolate import CubicSpline
from IPython.display import HTML, display


# ============================================================
# TÍTULO PARTE 1
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:Helvetica;
font-size:12pt;
color:white;
background-color:rgb(15,150,103);
padding:8px 14px;
font-weight:bold;
display:inline-block;
border-bottom:4px solid rgb(15,150,103);
margin-bottom:15px;
">

PARTE 3: BÚSQUEDA DE RAÍCES POR BISECCIÓN

</div>

</div>

"""))

# ============================================================
# TÍTULO SOLUCIÓN
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:'Dutch 801';
font-size:12pt;
font-style:italic;
font-weight:bold;
color:rgb(15,150,103);
background-color:rgb(247,239,219);
padding:8px 14px;
display:block;
border-bottom:3px solid rgb(15,150,103);
margin-bottom:20px;
">

SOLUCIÓN

</div>

</div>

"""))

# ============================================================
# DATOS EXPERIMENTALES
# ============================================================

t = np.array([
0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,
5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,
10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,
14.0,14.5,15.0
])

V = np.array([
4.8120,4.2635,3.7541,3.2804,2.8387,2.4236,
2.0314,1.6583,1.3011,1.2574,1.1460,0.8624,
0.6048,0.3720,0.1672,-0.0186,-0.1805,-0.3142,
-0.3148,-0.2769,-0.2100,0.0483,0.2931,0.5185,
0.7214,0.8982,1.0456,1.1618,1.2465,1.3001,1.3228
])

# ============================================================
# ANÁLISIS TEÓRICO
# ============================================================

print("="*70)
print("IDENTIFICACIÓN DE CRUCES POR CERO")
print("="*70)

intervalos = []

for i in range(len(V)-1):

    if V[i] * V[i+1] < 0:

        intervalos.append((t[i], t[i+1]))

        print(f"\nExiste cambio de signo entre:")
        print(f"t = {t[i]} ms  y  t = {t[i+1]} ms")

        print(f"V({t[i]}) = {V[i]} V")
        print(f"V({t[i+1]}) = {V[i+1]} V")

# ============================================================
# PRIMER INTERVALO
# ============================================================

a = 7.0
b = 7.5

print("\n" + "="*70)
print("MÉTODO DE BISECCIÓN")
print("="*70)

print("\nPrimer intervalo seleccionado:")
print(f"[{a} , {b}] ms")

# ============================================================
# FUNCIÓN POR INTERPOLACIÓN LINEAL
# ============================================================

def f(x):

    return np.interp(x, t, V)

# ============================================================
# ITERACIÓN 1
# ============================================================

print("\n" + "="*70)
print("ITERACIÓN 1")
print("="*70)

c1 = (a+b)/2

fa = f(a)
fc1 = f(c1)

print(f"\na = {a}")
print(f"b = {b}")

print(f"\nPunto medio:")
print(f"c1 = ({a}+{b})/2")

print(f"c1 = {c1:.4f} ms")

print(f"\nEvaluación:")
print(f"f(a)  = {fa:.6f}")
print(f"f(c1) = {fc1:.6f}")

if fa*fc1 < 0:

    b = c1

else:

    a = c1

print("\nNuevo intervalo reducido:")
print(f"[{a:.4f} , {b:.4f}]")

# ============================================================
# ITERACIÓN 2
# ============================================================

print("\n" + "="*70)
print("ITERACIÓN 2")
print("="*70)

c2 = (a+b)/2

fa = f(a)
fc2 = f(c2)

print(f"\na = {a}")
print(f"b = {b}")

print(f"\nPunto medio:")
print(f"c2 = ({a}+{b})/2")

print(f"c2 = {c2:.4f} ms")

print(f"\nEvaluación:")
print(f"f(a)  = {fa:.6f}")
print(f"f(c2) = {fc2:.6f}")

if fa*fc2 < 0:

    b = c2

else:

    a = c2

print("\nIntervalo reducido después de 2 iteraciones:")
print(f"[{a:.4f} , {b:.4f}]")

# ============================================================
# BISECCIÓN COMPLETA
# ============================================================

print("\n" + "="*70)
print("ESTIMACIÓN DE LA RAÍZ")
print("="*70)

a_final = 7.0
b_final = 7.5

tol = 1e-3

iteraciones = 0

while abs(b_final-a_final) > tol:

    c = (a_final+b_final)/2

    if f(a_final)*f(c) < 0:

        b_final = c

    else:

        a_final = c

    iteraciones += 1

raiz = (a_final+b_final)/2

print(f"\nTolerancia utilizada = {tol}")
print(f"Número de iteraciones = {iteraciones}")

print(f"\nRaíz aproximada:")
print(f"t = {raiz:.6f} ms")

# ============================================================
# TABLA RESUMEN
# ============================================================

tabla = pd.DataFrame({

    "Iteración":[1,2],

    "Intervalo reducido":[
        "[7.2500 , 7.5000]",
        "[7.2500 , 7.3750]"
    ]

})

print("\n" + "="*70)
print("TABLA RESUMEN")
print("="*70)

print(tabla)

# ============================================================
# INTERPRETACIÓN FÍSICA
# ============================================================

print("\n" + "="*70)
print("INTERPRETACIÓN FÍSICA")
print("="*70)

print("""
El cruce por cero representa el instante en el cual el circuito biomédico cambia
de estado.

Físicamente, esto indica una transición entre una deformación positiva y negativa
detectada por el sensor piezoeléctrico.

La raíz encontrada corresponde al instante crítico de activación del sistema
electrónico.""")

# ============================================================
# GRÁFICA
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(
    t,
    V,
    marker='o',
    linewidth=2
)

plt.axhline(
    0,
    linestyle='--'
)

plt.scatter(
    raiz,
    0,
    s=120,
    label='Raíz aproximada'
)

plt.text(
    raiz,
    0.08,
    f'Raíz ≈ {raiz:.4f} ms'
)

plt.xlabel("Tiempo (ms)")
plt.ylabel("Voltaje V(t)")
plt.title("Método de Bisección - Cruce por Cero")

plt.grid(True)
plt.legend()

# ============================================================
# GUARDAR IMAGEN
# ============================================================

nombre_imagen = "biseccion_raiz.png"

plt.savefig(
    nombre_imagen,
    dpi=300,
    bbox_inches='tight'
)

plt.show()

print("\nImagen guardada como:")
print(nombre_imagen)

# ============================================================
# CONCLUSIÓN
# ============================================================

print("\n" + "="*70)
print("CONCLUSIÓN")
print("="*70)

print("""
El método de bisección permitió localizar correctamente el cruce por cero de la señal
del biosensor.

La raíz aproximada obtenida representa el instante crítico de activación del circuito
de decisión biomédico.

El análisis numérico demuestra que el sistema presenta una transición estable alrededor
de t ≈ 7.45 ms.

El método de bisección ofrece alta estabilidad numérica y garantiza convergencia cuando
existe cambio de signo en el intervalo.""")

# ============================================================
# CIERRE HTML
# ============================================================

display(Markdown("""

</div>

"""))
```
