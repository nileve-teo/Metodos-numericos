El sistema requiere estimar con precisión el voltaje en los instantes:

<ul>

<li><b>t = 8.7 ms</b></li>
</ul>

Usando los datos de la tabla:

<ol>
<li>Construya un spline cúbico natural para toda la señal.</li>

<li>Estime V(4.3) y V(8.7) utilizando el spline cúbico natural.</li>

```python
# ============================================================
# MÉTODOS NUMÉRICOS semana08
# PARTE 1 : INTERPOLACIÓN CON SPLINE CÚBICO NATURAL
# ============================================================

# ============================================================
# LIBRERÍAS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# ============================================================
# CONFIGURACIÓN DE GRÁFICAS
# ============================================================

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 11

# ============================================================
# DATOS EXPERIMENTALES
# ============================================================

t = np.array([
0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,
5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,
10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,
14.0,14.5,15.0
])

V = np.array([
4.8120,4.2635,3.7541,3.2804,2.8387,2.4236,
2.0314,1.6583,1.3011,1.2574,1.1460,0.8624,
0.6048,0.3720,0.1672,-0.0186,-0.1805,-0.3142,
-0.3148,-0.2769,-0.2100,0.0483,0.2931,0.5185,
0.7214,0.8982,1.0456,1.1618,1.2465,1.3001,1.3228
])

# ============================================================
# TABLA DE DATOS EXPERIMENTALES
# ============================================================

tabla = pd.DataFrame({
    "Tiempo (ms)": t,
    "Voltaje V(t) [V]": V
})

print("================================================")
print("TABLA DE DATOS EXPERIMENTALES")
print("================================================\n")

print(tabla)

# ============================================================
# CONSTRUCCIÓN DEL SPLINE CÚBICO NATURAL
# ============================================================

spline = CubicSpline(t, V, bc_type='natural')

print("\n================================================")
print("SPLINE CÚBICO NATURAL CONSTRUIDO")
print("================================================\n")

print("El spline cúbico natural fue generado correctamente.")
print("Se garantiza continuidad de:")
print("• La función")
print("• La primera derivada")
print("• La segunda derivada")

# ============================================================
# PUNTOS A INTERPOLAR
# ============================================================

t1 = 4.3
t2 = 8.7

# ============================================================
# INTERPOLACIÓN MEDIANTE SPLINE
# ============================================================

V_t1 = float(spline(t1))
V_t2 = float(spline(t2))

# ============================================================
# INTERPOLACIÓN LINEAL SIMPLE
# ============================================================

V_t1_lineal = float(np.interp(t1, t, V))
V_t2_lineal = float(np.interp(t2, t, V))

# ============================================================
# RESULTADOS NUMÉRICOS
# ============================================================

print("\n================================================")
print("RESULTADOS MEDIANTE SPLINE CÚBICO NATURAL")
print("================================================\n")

print(f"V(4.3 ms) = {V_t1:.6f} V")
print(f"V(8.7 ms) = {V_t2:.6f} V")

print("\n================================================")
print("RESULTADOS MEDIANTE INTERPOLACIÓN LINEAL")
print("================================================\n")

print(f"V(4.3 ms) = {V_t1_lineal:.6f} V")
print(f"V(8.7 ms) = {V_t2_lineal:.6f} V")

# ============================================================
# TABLA COMPARATIVA
# ============================================================

comparacion = pd.DataFrame({
    "Punto": ["t = 4.3 ms", "t = 8.7 ms"],

    "Spline cúbico natural [V]":
        [round(V_t1,6),
         round(V_t2,6)],

    "Interpolación lineal [V]":
        [round(V_t1_lineal,6),
         round(V_t2_lineal,6)]
})

print("\n================================================")
print("TABLA COMPARATIVA")
print("================================================\n")

print(comparacion)

# ============================================================
# COMPARACIÓN CUALITATIVA
# ============================================================

print("\n================================================")
print("COMPARACIÓN CUALITATIVA")
print("================================================\n")

print("• El spline cúbico natural produce una curva suave y continua.")
print("• La interpolación lineal genera segmentos rectos.")
print("• El spline representa mejor el comportamiento físico del biosensor.")
print("• La interpolación lineal puede introducir cambios bruscos.")
print("• El spline ofrece mayor precisión en señales biomédicas.")
print("• El spline preserva continuidad en derivadas.")
print("• La interpolación lineal presenta cambios abruptos de pendiente.")

# ============================================================
# GENERACIÓN DE PUNTOS PARA LA CURVA
# ============================================================

tt = np.linspace(0, 15, 1000)
VV = spline(tt)

# ============================================================
# GRÁFICA PRINCIPAL
# ============================================================

plt.figure(figsize=(14,7))

plt.plot(
    tt,
    VV,
    linewidth=2.5,
    label='Spline cúbico natural'
)

plt.plot(
    t,
    V,
    'o',
    markersize=6,
    label='Datos experimentales'
)

plt.scatter(
    t1,
    V_t1,
    s=120,
    label='Interpolación en t = 4.3 ms'
)

plt.scatter(
    t2,
    V_t2,
    s=120,
    label='Interpolación en t = 8.7 ms'
)

plt.axhline(
    y=0,
    linestyle='--',
    linewidth=1
)

plt.xlabel('Tiempo t (ms)')
plt.ylabel('Voltaje V(t) [V]')

plt.title(
    'Interpolación mediante Spline Cúbico Natural',
    fontsize=15,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

plt.show()

# ============================================================
# GRÁFICA COMPARATIVA
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(
    tt,
    VV,
    linewidth=2.5,
    label='Spline cúbico natural'
)

plt.plot(
    t,
    V,
    '--',
    linewidth=1.5,
    label='Interpolación lineal'
)

plt.scatter(
    [t1,t2],
    [V_t1,V_t2],
    s=100,
    label='Valores spline'
)

plt.scatter(
    [t1,t2],
    [V_t1_lineal,V_t2_lineal],
    s=100,
    marker='x',
    linewidths=3,
    label='Valores lineales'
)

plt.xlabel('Tiempo t (ms)')
plt.ylabel('Voltaje V(t) [V]')

plt.title(
    'Comparación: Spline vs Interpolación Lineal',
    fontsize=15,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

plt.show()

# ============================================================
# CONCLUSIÓN
# ============================================================

print("\n================================================")
print("CONCLUSIÓN")
print("================================================\n")

print("La interpolación mediante spline cúbico natural")
print("permitió modelar de manera suave y continua")
print("la señal biomédica obtenida del sensor.")
print("Los resultados obtenidos representan mejor")
print("el comportamiento físico del biosensor en")
print("comparación con la interpolación lineal simple.")
```
