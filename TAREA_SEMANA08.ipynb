```python
# ============================================================
# MÉTODOS NUMÉRICOS - SEMANA 08
# PARTE 1: INTERPOLACIÓN CON SPLINE CÚBICO NATURAL
# ============================================================

# ============================================================
# LIBRERÍAS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
from IPython.display import HTML, display

# ============================================================
# CONFIGURACIÓN GENERAL
# ============================================================

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 11

# ============================================================
# TÍTULO PARTE 1
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:Helvetica;
font-size:12pt;
color:white;
background-color:rgb(15,150,103);
padding:8px 14px;
font-weight:bold;
display:inline-block;
border-bottom:4px solid rgb(15,150,103);
margin-bottom:15px;
">

PARTE 1: INTERPOLACIÓN CON SPLINE CÚBICO NATURAL

</div>

</div>

"""))

# ============================================================
# TÍTULO SOLUCIÓN
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:'Dutch 801';
font-size:12pt;
font-style:italic;
font-weight:bold;
color:rgb(15,150,103);
background-color:rgb(247,239,219);
padding:8px 14px;
display:block;
border-bottom:3px solid rgb(15,150,103);
margin-bottom:20px;
">

SOLUCIÓN

</div>

</div>

"""))

# ============================================================
# INTRODUCCIÓN
# ============================================================

print("La estimación precisa del voltaje biomédico")
print("requiere emplear métodos de interpolación")
print("que representen adecuadamente el comportamiento")
print("continuo de la señal fisiológica.\n")

print("Para ello se utilizará un spline cúbico natural,")
print("el cual garantiza continuidad en:")
print("• La función")
print("• La primera derivada")
print("• La segunda derivada\n")

print("Además, se compararán los resultados")
print("con una interpolación lineal simple.\n")


# ============================================================
# DATOS EXPERIMENTALES
# ============================================================

t = np.array([
0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,
5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,
10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,
14.0,14.5,15.0
])

V = np.array([
4.8120,4.2635,3.7541,3.2804,2.8387,2.4236,
2.0314,1.6583,1.3011,1.2574,1.1460,0.8624,
0.6048,0.3720,0.1672,-0.0186,-0.1805,-0.3142,
-0.3148,-0.2769,-0.2100,0.0483,0.2931,0.5185,
0.7214,0.8982,1.0456,1.1618,1.2465,1.3001,1.3228
])

# ============================================================
# TABLA DE DATOS
# ============================================================

tabla = pd.DataFrame({
    "Tiempo (ms)": t,
    "Voltaje V(t) [V]": V
})

print("================================================")
print("TABLA DE DATOS EXPERIMENTALES")
print("================================================\n")

print(tabla)

# ============================================================
# CONSTRUCCIÓN DEL SPLINE
# ============================================================

spline = CubicSpline(t, V, bc_type='natural')

print("\n================================================")
print("SPLINE CÚBICO NATURAL CONSTRUIDO")
print("================================================\n")

print("El spline cúbico natural fue generado correctamente.")
print("Se garantiza continuidad de:")
print("• La función")
print("• La primera derivada")
print("• La segunda derivada")

# ============================================================
# PUNTOS A INTERPOLAR
# ============================================================

t1 = 4.3
t2 = 8.7

# ============================================================
# INTERPOLACIÓN MEDIANTE SPLINE
# ============================================================

V_t1 = float(spline(t1))
V_t2 = float(spline(t2))

# ============================================================
# INTERPOLACIÓN LINEAL SIMPLE
# ============================================================

V_t1_lineal = float(np.interp(t1, t, V))
V_t2_lineal = float(np.interp(t2, t, V))

# ============================================================
# RESULTADOS SPLINE
# ============================================================

print("\n================================================")
print("RESULTADOS MEDIANTE SPLINE CÚBICO NATURAL")
print("================================================\n")

print(f"V(4.3 ms) = {V_t1:.6f} V")
print(f"V(8.7 ms) = {V_t2:.6f} V")

# ============================================================
# RESULTADOS LINEALES
# ============================================================

print("\n================================================")
print("RESULTADOS MEDIANTE INTERPOLACIÓN LINEAL")
print("================================================\n")

print(f"V(4.3 ms) = {V_t1_lineal:.6f} V")
print(f"V(8.7 ms) = {V_t2_lineal:.6f} V")

# ============================================================
# TABLA COMPARATIVA
# ============================================================

comparacion = pd.DataFrame({

    "Punto": [
        "t = 4.3 ms",
        "t = 8.7 ms"
    ],

    "Spline cúbico natural [V]": [
        round(V_t1,6),
        round(V_t2,6)
    ],

    "Interpolación lineal [V]": [
        round(V_t1_lineal,6),
        round(V_t2_lineal,6)
    ]
})

print("\n================================================")
print("TABLA COMPARATIVA")
print("================================================\n")

print(comparacion)

# ============================================================
# COMPARACIÓN CUALITATIVA
# ============================================================

print("\n================================================")
print("COMPARACIÓN CUALITATIVA")
print("================================================\n")

print("• El spline cúbico natural produce una curva suave y continua.")
print("• La interpolación lineal genera segmentos rectos.")
print("• El spline representa mejor el comportamiento físico del biosensor.")
print("• La interpolación lineal puede introducir cambios bruscos.")
print("• El spline ofrece mayor precisión en señales biomédicas.")
print("• El spline preserva continuidad en derivadas.")
print("• La interpolación lineal presenta cambios abruptos de pendiente.")

# ============================================================
# GENERACIÓN DE CURVA
# ============================================================

tt = np.linspace(0,15,1000)
VV = spline(tt)

# ============================================================
# GRÁFICA PRINCIPAL
# ============================================================

plt.figure(figsize=(14,7))

plt.plot(
    tt,
    VV,
    linewidth=2.5,
    label='Spline cúbico natural'
)

plt.plot(
    t,
    V,
    'o',
    markersize=6,
    label='Datos experimentales'
)

# ============================================================
# PUNTOS INTERPOLADOS
# ============================================================

plt.scatter(
    t1,
    V_t1,
    s=150,
    zorder=5,
    label='Punto 1'
)

plt.scatter(
    t2,
    V_t2,
    s=150,
    zorder=5,
    label='Punto 2'
)

# ============================================================
# ETIQUETAS DE LOS PUNTOS
# ============================================================

plt.annotate(
    '1',
    (t1, V_t1),
    textcoords="offset points",
    xytext=(10,10),
    fontsize=12,
    fontweight='bold'
)

plt.annotate(
    '2',
    (t2, V_t2),
    textcoords="offset points",
    xytext=(10,10),
    fontsize=12,
    fontweight='bold'
)

# ============================================================
# LÍNEA EN Y = 0
# ============================================================

plt.axhline(
    y=0,
    linestyle='--',
    linewidth=1
)

# ============================================================
# ETIQUETAS
# ============================================================

plt.xlabel('Tiempo t (ms)')
plt.ylabel('Voltaje V(t) [V]')

plt.title(
    'Interpolación mediante Spline Cúbico Natural',
    fontsize=15,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

plt.show()

# ============================================================
# GRÁFICA COMPARATIVA
# ============================================================

plt.figure(figsize=(13,6))

plt.plot(
    tt,
    VV,
    linewidth=2.5,
    label='Spline cúbico natural'
)

plt.plot(
    t,
    V,
    '--',
    linewidth=1.5,
    label='Interpolación lineal'
)

plt.scatter(
    [t1,t2],
    [V_t1,V_t2],
    s=120,
    label='Valores spline'
)

plt.scatter(
    [t1,t2],
    [V_t1_lineal,V_t2_lineal],
    s=120,
    marker='x',
    linewidths=3,
    label='Valores lineales'
)

plt.xlabel('Tiempo t (ms)')
plt.ylabel('Voltaje V(t) [V]')

plt.title(
    'Comparación: Spline vs Interpolación Lineal',
    fontsize=15,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

plt.show()

# ============================================================
# CONCLUSIÓN
# ============================================================

print("\n================================================")
print("CONCLUSIÓN")
print("================================================\n")

print("La interpolación mediante spline cúbico natural")
print("permitió modelar de manera suave y continua")
print("la señal biomédica obtenida del sensor.\n")

print("Los valores interpolados obtenidos fueron:")
print(f"• V(4.3 ms) = {V_t1:.6f} V")
print(f"• V(8.7 ms) = {V_t2:.6f} V\n")

print("El spline representó de mejor manera")
print("el comportamiento físico del biosensor")
print("comparado con la interpolación lineal.")
```

```python
# ============================================================
# PARTE 2 : DERIVACIÓN NUMÉRICA
# DIFERENCIAS CENTRADAS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# ============================================================
# CONTENEDOR GENERAL
# ============================================================

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 11

# ============================================================
# TÍTULO
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:Helvetica;
font-size:12pt;
color:white;
background-color:rgb(15,150,103);
padding:8px 14px;
font-weight:bold;
display:inline-block;
border-bottom:4px solid rgb(15,150,103);
margin-bottom:15px;
">

PARTE 2: DERIVACIÓN NUMÉRICA POR DIFERENCIAS CENTRADAS

</div>

</div>

"""))

#
# ============================================================
# SOLUCIÓN
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:'Dutch 801';
font-size:12pt;
font-style:italic;
font-weight:bold;
color:rgb(15,150,103);
background-color:rgb(247,239,219);
padding:8px 14px;
display:block;
border-bottom:3px solid rgb(15,150,103);
margin-bottom:20px;
">

SOLUCIÓN

</div>

</div>

"""))
# ============================================================
# DATOS
# ============================================================

t = np.array([
0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,
5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,
10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,
14.0,14.5,15.0
])

V = np.array([
4.8120,4.2635,3.7541,3.2804,2.8387,2.4236,
2.0314,1.6583,1.3011,1.2574,1.1460,0.8624,
0.6048,0.3720,0.1672,-0.0186,-0.1805,-0.3142,
-0.3148,-0.2769,-0.2100,0.0483,0.2931,0.5185,
0.7214,0.8982,1.0456,1.1618,1.2465,1.3001,1.3228
])

# ============================================================
# DIFERENCIAS CENTRADAS
# ============================================================

h = 0.5

def derivada_centrada(indice):
    return (V[indice+1] - V[indice-1])/(2*h)

# ============================================================
# ÍNDICES
# ============================================================

i4 = np.where(t == 4.0)[0][0]
i8 = np.where(t == 8.0)[0][0]
i12 = np.where(t == 12.0)[0][0]

# ============================================================
# DERIVADAS
# ============================================================

d4 = derivada_centrada(i4)
d8 = derivada_centrada(i8)
d12 = derivada_centrada(i12)

# ============================================================
# DESARROLLO TEÓRICO
# ============================================================

print("La tasa de cambio del voltaje del biosensor")
print("permite evaluar la respuesta dinámica del sistema.")
print("Para ello se emplea el método de diferencias centradas.\n")

print("FÓRMULA UTILIZADA:\n")

print("V'(t_i) = [V(t_i+h) - V(t_i-h)] / (2h)\n")

print(f"Tamaño del paso: h = {h} ms")
print(f"Entonces: 2h = {2*h} ms")

# ============================================================
# CÁLCULO 1
# ============================================================

print("\n" + "="*70)
print("1. CÁLCULO EN t = 4.0 ms")
print("="*70)

print("\nDatos vecinos:")

print(f"V(3.5 ms) = {V[i4-1]:.4f} V")
print(f"V(4.5 ms) = {V[i4+1]:.4f} V")

print("\nAplicando diferencias centradas:\n")

print("V'(4.0) = [V(4.5)-V(3.5)] / 1.0\n")

print(f"V'(4.0) = ({V[i4+1]:.4f} - {V[i4-1]:.4f}) / 1.0\n")

print(f"V'(4.0) = {d4:.4f} V/ms")

print("\nINTERPRETACIÓN FÍSICA:")

print("La derivada es negativa.")
print("El voltaje disminuye en esta región.")
print("Esto representa una relajación mecánica")
print("o disminución de la vibración detectada.")

# ============================================================
# CÁLCULO 2
# ============================================================

print("\n" + "="*70)
print("2. CÁLCULO EN t = 8.0 ms")
print("="*70)

print("\nDatos vecinos:")

print(f"V(7.5 ms) = {V[i8-1]:.4f} V")
print(f"V(8.5 ms) = {V[i8+1]:.4f} V")

print("\nAplicando diferencias centradas:\n")

print("V'(8.0) = [V(8.5)-V(7.5)] / 1.0\n")

print(f"V'(8.0) = ({V[i8+1]:.4f} - {V[i8-1]:.4f}) / 1.0\n")

print(f"V'(8.0) = {d8:.4f} V/ms")

print("\nINTERPRETACIÓN FÍSICA:")

print("La derivada es negativa.")
print("La señal continúa decreciendo")
print("cerca del valor mínimo del voltaje.")
print("Esto indica una posible transición")
print("del circuito biomédico.")

# ============================================================
# CÁLCULO 3
# ============================================================

print("\n" + "="*70)
print("3. CÁLCULO EN t = 12.0 ms")
print("="*70)

print("\nDatos vecinos:")

print(f"V(11.5 ms) = {V[i12-1]:.4f} V")
print(f"V(12.5 ms) = {V[i12+1]:.4f} V")

print("\nAplicando diferencias centradas:\n")

print("V'(12.0) = [V(12.5)-V(11.5)] / 1.0\n")

print(f"V'(12.0) = ({V[i12+1]:.4f} - {V[i12-1]:.4f}) / 1.0\n")

print(f"V'(12.0) = {d12:.4f} V/ms")

print("\nINTERPRETACIÓN FÍSICA:")

print("La derivada es positiva.")
print("El voltaje aumenta en esta región.")
print("Esto representa recuperación")
print("de la señal fisiológica.")
print("El tejido vuelve a deformarse")
print("generando mayor energía eléctrica.")

# ============================================================
# TABLA RESUMEN
# ============================================================

tabla = pd.DataFrame({

    "Tiempo (ms)":[4.0,8.0,12.0],

    "Derivada V'(t) [V/ms]":[
        round(float(d4),4),
        round(float(d8),4),
        round(float(d12),4)
    ],

    "Comportamiento":[
        "Decreciente",
        "Decreciente",
        "Creciente"
    ]
})

print("\n" + "="*70)
print("TABLA RESUMEN")
print("="*70 + "\n")

print(tabla)

# ============================================================
# GRÁFICA
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(
    t,
    V,
    marker='o',
    linewidth=2,
    label='Voltaje V(t)'
)

# Punto 1
plt.scatter(
    4.0,
    V[i4],
    s=150
)

plt.annotate(
    "1",
    (4.0, V[i4]),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

# Punto 2
plt.scatter(
    8.0,
    V[i8],
    s=150
)

plt.annotate(
    "2",
    (8.0, V[i8]),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

# Punto 3
plt.scatter(
    12.0,
    V[i12],
    s=150
)

plt.annotate(
    "3",
    (12.0, V[i12]),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

plt.axhline(
    0,
    linestyle='--'
)

plt.xlabel("Tiempo (ms)")
plt.ylabel("Voltaje V(t)")

plt.title(
    "Derivación Numérica por Diferencias Centradas",
    fontsize=14,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

# ============================================================
# GUARDAR IMAGEN
# ============================================================

nombre_imagen = "derivacion_numerica.png"

plt.savefig(
    nombre_imagen,
    dpi=300,
    bbox_inches='tight'
)

plt.show()

print("\nImagen guardada correctamente como:")
print(nombre_imagen)

# ============================================================
# MOSTRAR IMAGEN EN COLAB
# ============================================================

from IPython.display import Image, display

display(Image(filename=nombre_imagen))

# ============================================================
# CONCLUSIÓN
# ============================================================

print("\n" + "="*70)
print("CONCLUSIÓN")
print("="*70 + "\n")

print("El método de diferencias centradas permitió")
print("aproximar correctamente la derivada primera")
print("de la señal de voltaje del biosensor.")

print("\nLos resultados obtenidos muestran que:")

print("\n• En t = 4.0 ms la señal es decreciente,")
print("  indicando una disminución de la vibración.")

print("\n• En t = 8.0 ms el sistema continúa")
print("  acercándose a una zona crítica cercana")
print("  al mínimo de la señal.")

print("\n• En t = 12.0 ms la derivada positiva")
print("  representa recuperación del potencial")
print("  eléctrico detectado por el sensor.")

print("\nEl análisis numérico permite estudiar")
print("la dinámica del biosensor y evaluar")
print("el comportamiento fisiológico asociado")
print("a las microvibraciones del tejido.")

# ============================================================
# CIERRE HTML
# ============================================================

display(Markdown("""

</div>
</div>

"""))
```

```python
# ============================================================
# PARTE 3 : BÚSQUEDA DE RAÍCES POR BISECCIÓN
# SISTEMA DE TELEMETRÍA BIOMÉDICA
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# ============================================================
# CONTENEDOR GENERAL
# ============================================================
from scipy.interpolate import CubicSpline
from IPython.display import HTML, display


# ============================================================
# TÍTULO PARTE 1
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:Helvetica;
font-size:12pt;
color:white;
background-color:rgb(15,150,103);
padding:8px 14px;
font-weight:bold;
display:inline-block;
border-bottom:4px solid rgb(15,150,103);
margin-bottom:15px;
">

PARTE 3: BÚSQUEDA DE RAÍCES POR BISECCIÓN

</div>

</div>

"""))

# ============================================================
# TÍTULO SOLUCIÓN
# ============================================================

display(HTML("""

<div style="
max-width:21cm;
margin:auto;
">

<div style="
font-family:'Dutch 801';
font-size:12pt;
font-style:italic;
font-weight:bold;
color:rgb(15,150,103);
background-color:rgb(247,239,219);
padding:8px 14px;
display:block;
border-bottom:3px solid rgb(15,150,103);
margin-bottom:20px;
">

SOLUCIÓN

</div>

</div>

"""))

# ============================================================
# DATOS EXPERIMENTALES
# ============================================================

t = np.array([
0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,
5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,
10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,
14.0,14.5,15.0
])

V = np.array([
4.8120,4.2635,3.7541,3.2804,2.8387,2.4236,
2.0314,1.6583,1.3011,1.2574,1.1460,0.8624,
0.6048,0.3720,0.1672,-0.0186,-0.1805,-0.3142,
-0.3148,-0.2769,-0.2100,0.0483,0.2931,0.5185,
0.7214,0.8982,1.0456,1.1618,1.2465,1.3001,1.3228
])

# ============================================================
# IDENTIFICACIÓN DE CRUCES POR CERO
# ============================================================

intervalos = []

for i in range(len(V)-1):

    if V[i]*V[i+1] < 0:

        intervalos.append((t[i], t[i+1]))

# ============================================================
# MOSTRAR INTRODUCCIÓN
# ============================================================

print("El circuito de decisión cambia de estado")
print("cuando la señal del biosensor cruza")
print("el nivel cero.\n")

print("Para localizar el instante exacto")
print("del cruce se emplea el método")
print("de bisección.\n")

print("CRITERIO UTILIZADO:\n")

print("Existe una raíz en un intervalo [a,b]")
print("si la función cambia de signo:\n")

print("V(a) · V(b) < 0")

# ============================================================
# INTERVALOS CON CAMBIO DE SIGNO
# ============================================================

print("\n" + "="*70)
print("INTERVALOS DONDE OCURRE CRUCE POR CERO")
print("="*70 + "\n")

for k, inter in enumerate(intervalos):

    print(f"Intervalo {k+1}: [{inter[0]} ms , {inter[1]} ms]")

# ============================================================
# PRIMER INTERVALO
# ============================================================

a = 7.0
b = 7.5

fa = 0.1672
fb = -0.0186

print("\n" + "="*70)
print("PRIMER INTERVALO ANALIZADO")
print("="*70 + "\n")

print(f"a = {a} ms")
print(f"b = {b} ms")

print(f"\nV(a) = {fa} V")
print(f"V(b) = {fb} V")

print("\nComo:")
print("V(a) · V(b) < 0")

print("\nEntonces existe una raíz")
print("dentro del intervalo.")

# ============================================================
# ITERACIÓN 1
# ============================================================

m1 = (a+b)/2

# interpolación lineal aproximada
fm1 = fa + ((fb-fa)/(b-a))*(m1-a)

print("\n" + "="*70)
print("ITERACIÓN 1")
print("="*70 + "\n")

print(f"Punto medio m1 = ({a} + {b}) / 2")
print(f"m1 = {m1:.4f} ms")

print("\nAproximación del voltaje:")

print(f"V(m1) = {fm1:.6f} V")

if fa*fm1 < 0:

    b1 = m1
    a1 = a

else:

    a1 = m1
    b1 = b

print("\nNuevo intervalo reducido:")

print(f"[{a1:.4f} ms , {b1:.4f} ms]")

# ============================================================
# ITERACIÓN 2
# ============================================================

m2 = (a1+b1)/2

fm2 = fa + ((fb-fa)/(b-a))*(m2-a)

print("\n" + "="*70)
print("ITERACIÓN 2")
print("="*70 + "\n")

print(f"Punto medio m2 = ({a1:.4f} + {b1:.4f}) / 2")

print(f"m2 = {m2:.4f} ms")

print("\nAproximación del voltaje:")

print(f"V(m2) = {fm2:.6f} V")

if fa*fm2 < 0:

    b2 = m2
    a2 = a

else:

    a2 = m2
    b2 = b1

print("\nIntervalo reducido después")
print("de dos iteraciones:")

print(f"[{a2:.4f} ms , {b2:.4f} ms]")

# ============================================================
# BISECCIÓN COMPLETA
# ============================================================

tol = 1e-3

a_bis = 7.0
b_bis = 7.5

iteraciones = 0

while (b_bis-a_bis)/2 > tol:

    iteraciones += 1

    m = (a_bis+b_bis)/2

    fm = fa + ((fb-fa)/(b-a))*(m-a)

    if fa*fm < 0:

        b_bis = m

    else:

        a_bis = m

raiz = (a_bis+b_bis)/2

print("\n" + "="*70)
print("ESTIMACIÓN FINAL DE LA RAÍZ")
print("="*70 + "\n")

print(f"Tolerancia utilizada = {tol}")

print(f"\nNúmero de iteraciones = {iteraciones}")

print(f"\nRaíz aproximada:")

print(f"t = {raiz:.6f} ms")

# ============================================================
# TABLA RESUMEN
# ============================================================

tabla = pd.DataFrame({

    "Intervalo":[
        "[7.0 , 7.5]",
        "[10.0 , 10.5]"
    ],

    "Cambio de signo":[
        "Sí",
        "Sí"
    ]
})

print("\n" + "="*70)
print("TABLA RESUMEN")
print("="*70 + "\n")

print(tabla)

# ============================================================
# GRÁFICA
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(
    t,
    V,
    marker='o',
    linewidth=2,
    label='Voltaje V(t)'
)

plt.axhline(
    0,
    linestyle='--',
    linewidth=1.5,
    label='Nivel cero'
)

# raíz
plt.scatter(
    raiz,
    0,
    s=180,
    label='Raíz aproximada'
)

plt.annotate(
    "Raíz",
    (raiz,0),
    textcoords="offset points",
    xytext=(0,15),
    ha='center',
    fontsize=12,
    fontweight='bold'
)

plt.xlabel("Tiempo (ms)")
plt.ylabel("Voltaje V(t)")

plt.title(
    "Método de Bisección aplicado a la señal biomédica",
    fontsize=14,
    fontweight='bold'
)

plt.grid(True)
plt.legend()

# ============================================================
# GUARDAR IMAGEN
# ============================================================

nombre_imagen = "biseccion_raiz.png"

plt.savefig(
    nombre_imagen,
    dpi=300,
    bbox_inches='tight'
)

plt.show()

print("\nImagen guardada correctamente como:")
print(nombre_imagen)

# ============================================================
# MOSTRAR IMAGEN
# ============================================================

display(Image(filename=nombre_imagen))

# ============================================================
# CONCLUSIÓN
# ============================================================

print("\n" + "="*70)
print("CONCLUSIÓN")
print("="*70 + "\n")

print("El método de bisección permitió")
print("localizar correctamente el instante")
print("en el que la señal biomédica")
print("cruza el nivel cero.")

print("\nEl cruce identificado representa")
print("una posible transición crítica")
print("en el circuito de decisión del")
print("sistema de telemetría biomédica.")

print("\nLa aproximación obtenida con")
print("tolerancia de 10^-3 ms garantiza")
print("una estimación numérica precisa")
print("del instante de activación.")

# ============================================================
# CIERRE HTML
# ============================================================

display(Markdown("""

</div>
</div>

"""))
```
