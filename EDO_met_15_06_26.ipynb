<p>
En la ingeniería estructural de rascacielos frente a sismos, se utilizan disipadores viscosos colocados en las uniones de las vigas. Cuando el edificio se mueve, el fluido dentro del amortiguador genera una fuerza de oposición.
</p>

<p>
Para movimientos severos, la resistencia del fluido deja de ser lineal y se modela mediante la ecuación de fricción cuadrática de Van der Pol:
</p>

<div align="center">

$$
\frac{d^2x}{dt^2}-\mu(1-x^2)\frac{dx}{dt}+x=0
$$

</div>

<p>
Donde \(x(t)\) representa el desplazamiento normalizado de la estructura y \(\mu = 1.5\) es el coeficiente de amortiguamiento no lineal.
</p>

<p>
Suponga que una réplica sísmica inicial desplaza la estructura de su eje central, dejando las condiciones iniciales:
</p>

<div align="center">

$$
x(0)=2   
$$

$$
\dot{x}(0)=0   
$$

</div>
<p>
Siendo 2 el desplazamiento   máximo y 0 el reposo instantáneo.
</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

# =====================================================
# CONFIGURACIÓN GRÁFICA GLOBAL
# =====================================================

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['figure.dpi'] = 120

# =====================================================
# PARÁMETROS
# =====================================================

mu = 1.5
t0 = 0
tf = 20
h = 0.05

# =====================================================
# FUNCIÓN VAN DER POL
# =====================================================

def f(x, v):
    dxdt = v
    dvdt = mu*(1 - x**2)*v - x
    return dxdt, dvdt

# =====================================================
# FUNCIÓN RK4
# =====================================================

def RK4(h):

    N = int((tf-t0)/h)

    t = np.linspace(t0, tf, N+1)

    x = np.zeros(N+1)
    v = np.zeros(N+1)

    x[0] = 2.0
    v[0] = 0.0

    for i in range(N):

        k1x, k1v = f(x[i],v[i])

        k2x, k2v = f(
            x[i] + h*k1x/2,
            v[i] + h*k1v/2
        )

        k3x, k3v = f(
            x[i] + h*k2x/2,
            v[i] + h*k2v/2
        )

        k4x, k4v = f(
            x[i] + h*k3x,
            v[i] + h*k3v
        )

        x[i+1] = x[i] + h*(k1x + 2*k2x + 2*k3x + k4x)/6
        v[i+1] = v[i] + h*(k1v + 2*k2v + 2*k3v + k4v)/6

    return t,x,v

# =====================================================
# SOLUCIÓN PRINCIPAL
# =====================================================

t,x,v = RK4(0.05)

# aceleración
a = mu*(1-x**2)*v - x

# =====================================================
# TABLA PRIMERAS 10 ITERACIONES
# =====================================================

tabla10 = pd.DataFrame({
    "Iter": np.arange(10),
    "t [s]": np.round(t[:10],4),
    "x(t)": np.round(x[:10],6),
    "v(t)": np.round(v[:10],6)
})

print("\n"+"="*70)
print("PRIMERAS 10 ITERACIONES")
print("="*70)
print(tabla10.to_string(index=False))

# =====================================================
# TABLA ESTADOS FINALES
# =====================================================

tabla_final = pd.DataFrame({
    "Variable":["Tiempo final","Desplazamiento final","Velocidad final"],
    "Valor":[
        round(t[-1],4),
        round(x[-1],6),
        round(v[-1],6)
    ]
})

print("\n"+"="*70)
print("ESTADO FINAL")
print("="*70)
print(tabla_final.to_string(index=False))

# =====================================================
# FIGURA 1
# EVOLUCIÓN TEMPORAL
# =====================================================

fig, ax1 = plt.subplots(figsize=(13,6))

ax1.plot(
    t,
    x,
    color='#1f77b4',
    linewidth=3,
    label='Desplazamiento x(t)'
)

ax1.plot(
    t,
    v,
    color='#d62728',
    linewidth=3,
    linestyle='--',
    label='Velocidad v(t)'
)

idx = np.argmax(np.abs(a))

ax1.axvspan(
    max(0,t[idx]-0.5),
    min(tf,t[idx]+0.5),
    color='gold',
    alpha=0.25,
    label='Zona de máxima aceleración'
)

ax1.scatter(
    t[idx],
    x[idx],
    s=120,
    color='black',
    zorder=5
)

ax1.annotate(
    f'Máxima |a| = {abs(a[idx]):.2f}',
    xy=(t[idx],x[idx]),
    xytext=(t[idx]+1.2,x[idx]+1),
    fontsize=11,
    arrowprops=dict(
        arrowstyle='->',
        lw=1.5
    )
)

ax1.set_title(
    'Evolución Temporal del Oscilador de Van der Pol',
    fontweight='bold',
    pad=15
)

ax1.set_xlabel('Tiempo [s]', fontweight='bold')
ax1.set_ylabel('Magnitud', fontweight='bold')

ax1.grid(
    True,
    linestyle='--',
    alpha=0.4
)

ax1.minorticks_on()

ax1.legend(
    loc='best',
    frameon=True
)

plt.tight_layout()
plt.show()

# =====================================================
# FIGURA 2
# ESPACIO DE FASES MEJORADO
# =====================================================

fig, ax = plt.subplots(figsize=(9,8))

points = np.array([x,v]).T.reshape(-1,1,2)
segments = np.concatenate(
    [points[:-1],points[1:]],
    axis=1
)

lc = LineCollection(
    segments,
    cmap='plasma',
    linewidth=3
)

lc.set_array(t)

ax.add_collection(lc)

ax.scatter(
    x[0],
    v[0],
    s=180,
    marker='o',
    color='green',
    label='Inicio'
)

ax.scatter(
    x[-1],
    v[-1],
    s=180,
    marker='s',
    color='red',
    label='Final'
)

for i in range(0,len(x)-40,40):

    ax.arrow(
        x[i],
        v[i],
        x[i+1]-x[i],
        v[i+1]-v[i],
        head_width=0.08,
        alpha=0.5
    )

ax.set_xlim(min(x)-0.5,max(x)+0.5)
ax.set_ylim(min(v)-0.5,max(v)+0.5)

ax.set_xlabel(
    'Desplazamiento x',
    fontweight='bold'
)

ax.set_ylabel(
    'Velocidad v',
    fontweight='bold'
)

ax.set_title(
    'Diagrama de Espacio de Fases',
    fontweight='bold',
    pad=15
)

ax.grid(
    True,
    linestyle='--',
    alpha=0.4
)

ax.legend()

cbar = plt.colorbar(lc)
cbar.set_label(
    'Tiempo [s]',
    fontsize=12,
    fontweight='bold'
)

plt.tight_layout()
plt.show()

 # =====================================================
# FIGURA 3
# PRUEBA DE ESTABILIDAD NUMÉRICA
# =====================================================

fig, axs = plt.subplots(
    3,1,
    figsize=(13,10),
    sharex=True
)

colores = [
    '#1f77b4',
    '#ff7f0e',
    '#d62728'
]

axs[0].plot(
    t1,x1,
    color=colores[0],
    linewidth=3
)

axs[1].plot(
    t2,x2,
    color=colores[1],
    linewidth=3
)

axs[2].plot(
    t3,x3,
    color=colores[2],
    linewidth=3
)

titulos = [
    'h = 0.05 s (Solución Estable)',
    'h = 0.50 s (Error Creciente)',
    'h = 1.00 s (Divergencia Numérica)'
]

for i in range(3):

    axs[i].set_title(
        titulos[i],
        fontsize=13,
        fontweight='bold'
    )

    axs[i].set_ylabel(
        'x(t)',
        fontweight='bold'
    )

    axs[i].grid(
        True,
        linestyle='--',
        alpha=0.4
    )

axs[2].set_yscale('symlog')

axs[2].set_xlabel(
    'Tiempo [s]',
    fontweight='bold'
)

fig.suptitle(
    'Comparación de Estabilidad Numérica del Método RK4',
    fontsize=18,
    fontweight='bold',
    y=0.98
)

plt.tight_layout()
plt.show()

# ==========================================
# PRUEBA DE ESTABILIDAD
# ==========================================

t1, x1, v1 = RK4(0.05)

with np.errstate(over='ignore', invalid='ignore'):
    t2, x2, v2 = RK4(0.5)
    t3, x3, v3 = RK4(1.0)

# =====================================================
# TABLA DE DIVERGENCIA
# =====================================================

print("\n")
print("="*70)
print("PRUEBA DE ESTABILIDAD NUMÉRICA")
print("="*70)

print(f"h = 0.05  --> x_final = {x1[-1]:.6f}")
print(f"h = 0.50  --> x_final = {x2[-1]:.6f}")
print(f"h = 1.00  --> x_final = {x3[-1]:.6f}")
```

## Q1. Análisis de la Evolución Temporal

Al analizar la **Gráfica 1**, se observa que las mayores aceleraciones ocurren durante el intervalo inicial del movimiento, aproximadamente entre **0 s y 2 s**. En esta región el sistema parte de una condición inicial alejada del equilibrio ((x(0)=2,;v(0)=0)), por lo que la fuerza restauradora y el término no lineal del oscilador generan cambios rápidos en la velocidad.

Estas aceleraciones elevadas se reflejan en las **pendientes pronunciadas** de las curvas de desplazamiento (x(t)) y velocidad (v(t)). A medida que transcurre el tiempo, la amplitud de las variaciones disminuye y las pendientes se vuelven más suaves, indicando que el sistema se aproxima gradualmente a un régimen oscilatorio estable.

La zona señalada en la gráfica corresponde al instante donde el valor absoluto de la aceleración

$$
a(t)=\mu(1-x^2)v-x
$$

alcanza su máximo valor, evidenciando el periodo de mayor intercambio de energía dentro del sistema dinámico.

## Q2. Interpretación Física del Diagrama de Fases

En la **Gráfica 2** se observa que la trayectoria en el espacio de fases no converge hacia el origen ((0,0)), sino que evoluciona hasta alcanzar una **curva cerrada estable**, conocida como **ciclo límite**.

Físicamente, esto significa que la estructura no pierde completamente su movimiento debido al amortiguamiento. El término no lineal presente en la ecuación de Van der Pol actúa de manera diferente según la amplitud de la oscilación:

* Para amplitudes pequeñas, el sistema introduce energía al movimiento.
* Para amplitudes grandes, el sistema disipa energía.

Como consecuencia, se establece un equilibrio dinámico entre la energía suministrada y la energía disipada. Este balance impide que la oscilación desaparezca completamente y provoca que la estructura mantenga una oscilación periódica de amplitud prácticamente constante.

La convergencia de diferentes trayectorias hacia la misma órbita cerrada indica además que el comportamiento final del sistema es independiente de las condiciones iniciales y está gobernado por la dinámica propia del ciclo límite.

## Q3. Prueba de Estabilidad Numérica

Para evaluar la estabilidad del método RK4, se incrementó el tamaño de paso desde el valor original (h=0.05,s) hasta (h=0.5,s) y posteriormente hasta (h=1.0,s).

### Caso 1: (h=0.5,s)

Con un tamaño de paso de (0.5,s), la solución todavía conserva la forma general de la respuesta física, aunque aparecen deformaciones apreciables en la amplitud y en la frecuencia de las oscilaciones. Esto indica que el error numérico comienza a acumularse y la aproximación pierde precisión.

### Caso 2: (h=1.0,s)

Cuando el tamaño de paso aumenta a (1.0,s), la solución se vuelve inestable y aparece un fenómeno de **divergencia numérica**. Los valores calculados crecen rápidamente hasta alcanzar magnitudes extremadamente grandes, del orden de

$$
10^{287}
$$

produciendo advertencias de *overflow* y valores numéricamente inválidos.

En la gráfica comparativa se observa que la trayectoria deja de representar el comportamiento físico del sistema y se desvía violentamente de la solución correcta.

### Justificación del fallo del método RK4

Aunque RK4 es un método de cuarto orden con alta precisión, su estabilidad depende del tamaño de paso utilizado. La ecuación de Van der Pol contiene términos altamente no lineales, particularmente

$$
\mu(1-x^2)v
$$

los cuales pueden variar rápidamente en intervalos cortos de tiempo.

Al utilizar un paso excesivamente grande, el método no logra capturar adecuadamente dichas variaciones y el error local de truncamiento se amplifica en cada iteración. Como resultado:

1. Se generan errores crecientes en el cálculo de (x) y (v).
2. Los términos no lineales amplifican dichos errores.
3. Los valores calculados exceden la capacidad numérica del tipo de dato utilizado.
4. La solución diverge y pierde completamente su significado físico.

Por lo tanto, para este problema se concluye que un tamaño de paso de (h=0.05,s) proporciona una solución estable y confiable, mientras que valores de (h=0.5,s) y especialmente (h=1.0,s) comprometen severamente la estabilidad y precisión del método RK4.
