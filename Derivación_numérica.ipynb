Se desarrolló un programa en Python para calcular numéricamente las derivadas de la función $f(x)=e^x$ mediante métodos de diferencias finitas.

El algoritmo evalúa distintos tamaños de paso \(h\) y determina el error porcentual asociado a cada aproximación, permitiendo comparar la precisión de los resultados obtenidos.

```python
import math
import pandas as pd

# Función
def f(x):
    return math.exp(x)

# Valor exacto
valor_exacto = math.exp(2)

# Valores de h
h_values = [0.5, 0.1, 0.05, 0.01, 0.005, 0.001]

# Listas para almacenar resultados
tabla = []

print("="*130)
print("DERIVACIÓN NUMÉRICA DE f(x)=e^x")
print("="*130)

for h in h_values:

    # Primera derivada (diferencia centrada)
    f1 = (f(2+h) - f(2-h)) / (2*h)

    # Segunda derivada
    f2 = (f(2+h) - 2*f(2) + f(2-h)) / (h**2)

    # Aproximaciones con 6 y 8 decimales
    f1_6 = round(f1, 6)
    f1_8 = round(f1, 8)

    f2_6 = round(f2, 6)
    f2_8 = round(f2, 8)

    # Error porcentual
    error1_6 = abs((valor_exacto - f1_6)/valor_exacto)*100
    error1_8 = abs((valor_exacto - f1_8)/valor_exacto)*100

    error2_6 = abs((valor_exacto - f2_6)/valor_exacto)*100
    error2_8 = abs((valor_exacto - f2_8)/valor_exacto)*100

    tabla.append([
        h,
        f1_6,
        f1_8,
        error1_6,
        error1_8,
        f2_6,
        f2_8,
        error2_6,
        error2_8
    ])

# Crear DataFrame
df = pd.DataFrame(tabla, columns=[
    "h",
    "f'(2) 6 dec",
    "f'(2) 8 dec",
    "% Error 6 dec f'",
    "% Error 8 dec f'",
    'f"(2) 6 dec',
    'f"(2) 8 dec',
    '% Error 6 dec f"',
    '% Error 8 dec f"'
])

# Mostrar tabla
print(df.to_string(index=False))
```

# Análisis y comparación de resultados

Al comparar los resultados obtenidos mediante la solución analítica y la solución mediante  Python, se observa que ambas aproximaciones presentan una tendencia consistente: el error disminuye conforme el valor del paso \(h\) se hace más pequeño.

Sin embargo, existen diferencias numéricas entre ambas tablas debido al criterio utilizado para calcular el error.

En la primera tabla se empleó el error absoluto, definido como:

$$
E_a = |V_{exacto}-V_{aproximado}|
$$

mientras que en la solución mediante Python se utilizó el error porcentual:

$$
E_{\%}=
\left|
\frac{V_{exacto}-V_{aproximado}}
{V_{exacto}}
\right|\times100
$$

Por esta razón, los valores numéricos no coinciden exactamente, aunque ambos procedimientos son matemáticamente correctos y válidos dentro del análisis numérico.

Además, se aprecia que para valores grandes de \(h\), como \(0.5\), el error es considerablemente mayor, ya que la aproximación mediante diferencias finitas pierde precisión al utilizar intervalos amplios.

En cambio, cuando \(h\) disminuye hasta valores como \(0.01\), \(0.005\) y \(0.001\), las aproximaciones obtenidas para $f'(2)$ y $f''(2)$ convergen al valor exacto:

$$
e^2 = 7.3890560989
$$

Esto demuestra que el método numérico empleado presenta convergencia y estabilidad para tamaños de paso pequeños.

Por lo tanto, tanto la solución mediante Phyton como la solución computacional validan el correcto funcionamiento de las fórmulas de diferencias centradas y diferencias finitas utilizadas en el desarrollo del ejercicio.

# **a.Método de Diferencias Finitas Centradas**

A continuación, se presenta la implementación desarrollada en Python para aproximar la primera y segunda derivada de la función a partir de datos tabulados.

El programa emplea métodos de diferencias finitas y diferencias centradas para calcular  $f'(2.25)$ y $f''(2.25)$, utilizando los puntos más cercanos al valor de evaluación.

Además, el código muestra paso a paso el procedimiento realizado, incluyendo la selección de puntos, sustitución de valores y resultados finales obtenidos numéricamente.

```python
import numpy as np
import pandas as pd

# =========================================================
# DERIVACIÓN NUMÉRICA
# =========================================================

print("="*80)
print("DERIVACIÓN NUMÉRICA")
print("="*80)

# Datos tabulados
x = np.array([1.5, 1.9, 2.1, 2.4, 2.6, 3.1])
fx = np.array([1.0628, 1.3961, 1.5432, 1.7349, 1.8423, 2.0397])

# Mostrar tabla de datos
tabla_datos = pd.DataFrame({
    "x": x,
    "f(x)": fx
})

print("\nDATOS TABULADOS:\n")
print(tabla_datos.to_string(index=False))

# =========================================================
# PUNTO DE EVALUACIÓN
# =========================================================

x_eval = 2.25

print("\n")
print("="*80)
print(f"Cálculo de f'(2.25) y f''(2.25)")
print("="*80)

# =========================================================
# SELECCIÓN DE PUNTOS CERCANOS
# =========================================================

x0 = 2.1
x1 = 2.4

f_x0 = 1.5432
f_x1 = 1.7349

# Distancia entre puntos
h = x1 - x0

print(f"\nPuntos utilizados:")
print(f"x0 = {x0}     f(x0) = {f_x0}")
print(f"x1 = {x1}     f(x1) = {f_x1}")

print(f"\nPaso h = {h}")

# =========================================================
# PRIMERA DERIVADA
# =========================================================

print("\n")
print("="*80)
print("PRIMERA DERIVADA")
print("="*80)

# Diferencia centrada
f1 = (f_x1 - f_x0) / h

print("\nFórmula utilizada:")
print("f'(x) ≈ [f(x+h)-f(x-h)] / 2h")

print("\nSustituyendo valores:")
print(f"f'(2.25) ≈ ({f_x1} - {f_x0}) / {h}")

print(f"\nf'(2.25) ≈ {f1:.6f}")

# =========================================================
# SEGUNDA DERIVADA
# =========================================================

print("\n")
print("="*80)
print("SEGUNDA DERIVADA")
print("="*80)

# Aproximación de f(2.25)
f_centro = (f_x0 + f_x1) / 2

# h/2 porque 2.25 está en el centro
h2 = 0.15

# Segunda derivada
f2 = (f_x1 - 2*f_centro + f_x0) / (h2**2)

print("\nFórmula utilizada:")
print("f''(x) ≈ [f(x+h)-2f(x)+f(x-h)] / h²")

print("\nAproximación del valor central:")
print(f"f(2.25) ≈ ({f_x0} + {f_x1}) / 2")

print(f"f(2.25) ≈ {f_centro:.6f}")

print("\nSustituyendo valores:")
print(f"f''(2.25) ≈ ({f_x1} - 2({f_centro}) + {f_x0}) / ({h2}²)")

print(f"\nf''(2.25) ≈ {f2:.6f}")

# =========================================================
# RESULTADOS FINALES
# =========================================================

print("\n")
print("="*80)
print("RESULTADOS FINALES")
print("="*80)

print(f"\nf'(2.25)  ≈  {f1:.6f}")
print(f"f''(2.25) ≈  {f2:.6f}")
```

# **b.Método de Interpolación Polinómica de Lagrange**


**Interpolación  polinómica de grado 5**

Con el objetivo de obtener una aproximación más precisa de la función a partir de los datos tabulados, se implementó un polinomio interpolante de grado 5 utilizando el método de Lagrange.

El programa desarrollado en Python construye automáticamente el polinomio interpolante que pasa por todos los puntos dados y posteriormente evalúa:

$$
f(2.25), \quad f'(2.25), \quad \text{y} \quad f''(2.25)
$$

mediante derivación simbólica del polinomio obtenido.

Este procedimiento permite aproximar las derivadas de la función utilizando interpolación polinómica y comparar los resultados con los obtenidos mediante diferencias finitas.

```python
import numpy as np
import sympy as sp
import pandas as pd

# =========================================================
# INTERPOLACIÓN POLINÓMICA DE GRADO 5
# =========================================================

print("="*160)
print("POLINOMIO INTERPOLANTE DE GRADO 5")
print("="*160)

# Datos tabulados
x_datos = np.array([1.5, 1.9, 2.1, 2.4, 2.6, 3.1])
y_datos = np.array([1.0628, 1.3961, 1.5432, 1.7349, 1.8423, 2.0397])

# Mostrar tabla
tabla = pd.DataFrame({
    "x": x_datos,
    "f(x)": y_datos
})

print("\nDATOS TABULADOS:\n")
print(tabla.to_string(index=False))

# =========================================================
# VARIABLE SIMBÓLICA
# =========================================================

x = sp.Symbol('x')

# =========================================================
# POLINOMIO DE LAGRANGE
# =========================================================

P = 0

n = len(x_datos)

for i in range(n):

    # Término de Lagrange
    Li = 1

    for j in range(n):

        if i != j:
            Li *= (x - x_datos[j]) / (x_datos[i] - x_datos[j])

    # Construcción del polinomio
    P += y_datos[i] * Li

# Expandir polinomio
P_expandido = sp.expand(P)

print("\n")
print("="*160)
print("POLINOMIO INTERPOLANTE")
print("="*160)

print("\nP(x) =")
print(sp.expand(P_expandido))

# =========================================================
# EVALUACIÓN EN x = 2.25
# =========================================================

x_eval = 2.25

valor = P_expandido.subs(x, x_eval)

print("\n")
print("="*160)
print("EVALUACIÓN DEL POLINOMIO")
print("="*160)

print(f"\nP(2.25) = {float(valor):.8f}")

# =========================================================
# PRIMERA DERIVADA
# =========================================================

P1 = sp.diff(P_expandido, x)

valor1 = P1.subs(x, x_eval)

print("\n")
print("="*160)
print("PRIMERA DERIVADA")
print("="*160)

print("\nP'(x) =")
print(sp.expand(P1))

print(f"\nP'(2.25) = {float(valor1):.8f}")

# =========================================================
# SEGUNDA DERIVADA
# =========================================================

P2 = sp.diff(P1, x)

valor2 = P2.subs(x, x_eval)

print("\n")
print("="*160)
print("SEGUNDA DERIVADA")
print("="*160)

print("\nP''(x) =")
print(sp.expand(P2))

print(f"\nP''(2.25) = {float(valor2):.8f}")

# =========================================================
# RESULTADOS FINALES
# =========================================================

print("\n")
print("="*160)
print("RESULTADOS FINALES")
print("="*160)

print(f"\nf(2.25)    ≈ {float(valor):.8f}")
print(f"f'(2.25)   ≈ {float(valor1):.8f}")
print(f"f''(2.25)  ≈ {float(valor2):.8f}")
```

# Análisis de resultados

Mediante la interpolación polinómica de grado 5 se obtuvo un polinomio capaz de aproximar el comportamiento de la función a partir de los datos tabulados proporcionados.

La evaluación del polinomio en \(x=2.25\) permitió obtener:

$$
f(2.25)\approx1.64355767
$$

Asimismo, al derivar simbólicamente el polinomio interpolante se calcularon la primera y segunda derivada:

$$

$$

y

$$
f''(2.25)\approx-0.40135787
$$

El valor positivo de la primera derivada indica que la función presenta crecimiento alrededor de \(x=2.25\).

Por otro lado, la segunda derivada negativa sugiere que la curva presenta concavidad hacia abajo en dicho intervalo.

Además, se observa que el valor de la primera derivada es muy cercano al obtenido mediante diferencias finitas, lo cual valida la consistencia de ambos métodos numéricos.

Las pequeñas diferencias entre resultados se deben a que la interpolación polinómica utiliza todos los puntos tabulados para construir una aproximación global de la función, mientras que las diferencias finitas emplean únicamente puntos cercanos al valor de evaluación.

# **c. Método de Splines Cúbicos**

Con el propósito de obtener una aproximación más suave y precisa de la función a partir de los datos tabulados, se aplicó el método de Splines Cúbicos.

Este método construye polinomios cúbicos por intervalos, garantizando continuidad en la función y en sus derivadas primeras y segundas.

Mediante la implementación en Python se evaluaron:

$$
f(2.25), \quad f'(2.25), \quad \text{y} \quad f''(2.25)
$$

utilizando interpolación spline cúbica sobre los datos proporcionados.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# =========================================================
# SPLINES CÚBICOS
# =========================================================

print("="*120)
print("MÉTODO DE SPLINES CÚBICOS")
print("="*120)

# Datos tabulados
x = np.array([1.5, 1.9, 2.1, 2.4, 2.6, 3.1])
fx = np.array([1.0628, 1.3961, 1.5432, 1.7349, 1.8423, 2.0397])

# Mostrar datos
tabla = pd.DataFrame({
    "x": x,
    "f(x)": fx
})

print("\nDATOS TABULADOS:\n")
print(tabla.to_string(index=False))

# =========================================================
# CONSTRUCCIÓN DEL SPLINE CÚBICO
# =========================================================

spline = CubicSpline(x, fx)

print("\n")
print("="*120)
print("EVALUACIÓN EN x = 2.25")
print("="*120)

x_eval = 2.25

# Evaluación de la función
f_eval = spline(x_eval)

# Primera derivada
f1 = spline.derivative(1)(x_eval)

# Segunda derivada
f2 = spline.derivative(2)(x_eval)

# =========================================================
# RESULTADOS
# =========================================================

print(f"\nf(2.25)    ≈ {f_eval:.8f}")
print(f"f'(2.25)   ≈ {f1:.8f}")
print(f"f''(2.25)  ≈ {f2:.8f}")

# =========================================================
# ECUACIONES DE LOS SPLINES
# =========================================================

print("\n")
print("="*120)
print("POLINOMIOS SPLINE POR INTERVALOS")
print("="*120)

for i in range(len(spline.c[0])):

    a = spline.c[0][i]
    b = spline.c[1][i]
    c = spline.c[2][i]
    d = spline.c[3][i]

    print(f"\nIntervalo [{x[i]}, {x[i+1]}]")

    print(f"S(x) = "
          f"{a:.6f}(x-{x[i]})³ + "
          f"{b:.6f}(x-{x[i]})² + "
          f"{c:.6f}(x-{x[i]}) + "
          f"{d:.6f}")

# =========================================================
# GRÁFICA DEL SPLINE
# =========================================================

# Valores suaves para la curva
x_suave = np.linspace(min(x), max(x), 500)

# Evaluación spline
y_suave = spline(x_suave)

plt.figure(figsize=(10,6))

# Puntos originales
plt.plot(x, fx, 'o', markersize=8, label='Datos tabulados')

# Curva spline
plt.plot(x_suave, y_suave, linewidth=2, label='Spline cúbico')

# Punto evaluado
plt.plot(x_eval, f_eval, 's', markersize=8,
         label=f'Evaluación en x={x_eval}')

# Configuración
plt.title("Interpolación mediante Splines Cúbicos",
          fontsize=16)

plt.xlabel("x", fontsize=13)
plt.ylabel("f(x)", fontsize=13)

plt.grid(True)
plt.legend()

plt.show()

# =========================================================
# RESULTADOS FINALES
# =========================================================

print("\n")
print("="*120)
print("RESULTADOS FINALES")
print("="*120)

print(f"\nf(2.25)    ≈ {f_eval:.8f}")
print(f"f'(2.25)   ≈ {f1:.8f}")
print(f"f''(2.25)  ≈ {f2:.8f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import sympy as sp

# =========================================================
# DATOS TABULADOS
# =========================================================

x = np.array([1.5, 1.9, 2.1, 2.4, 2.6, 3.1])
fx = np.array([1.0628, 1.3961, 1.5432, 1.7349, 1.8423, 2.0397])

x_eval = 2.25

# =========================================================
# DIFERENCIAS FINITAS
# =========================================================

# Aproximación local
f1_df = (1.7349 - 1.5432) / 0.3

# =========================================================
# LAGRANGE
# =========================================================

X = sp.Symbol('X')

P = 0
n = len(x)

for i in range(n):

    Li = 1

    for j in range(n):

        if i != j:
            Li *= (X - x[j]) / (x[i] - x[j])

    P += fx[i] * Li

P_expandido = sp.expand(P)

# Función numérica
P_func = sp.lambdify(X, P_expandido, 'numpy')

# Derivada
P1 = sp.diff(P_expandido, X)
P1_func = sp.lambdify(X, P1, 'numpy')

# =========================================================
# SPLINE CÚBICO
# =========================================================

spline = CubicSpline(x, fx)

# =========================================================
# PUNTOS SUAVES
# =========================================================

x_suave = np.linspace(min(x), max(x), 500)

y_lagrange = P_func(x_suave)
y_spline = spline(x_suave)

# =========================================================
# GRÁFICA 1
# DIFERENCIAS FINITAS
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(x, fx, 'o-', linewidth=2)

plt.plot([2.1, 2.4],
         [1.5432, 1.7349],
         linewidth=3)

plt.title("Método de Diferencias Finitas",
          fontsize=15)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True)

plt.show()

# =========================================================
# GRÁFICA 2
# LAGRANGE
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(x, fx, 'o',
         markersize=8,
         label='Datos tabulados')

plt.plot(x_suave, y_lagrange,
         linewidth=2,
         label='Polinomio de Lagrange')

plt.plot(x_eval,
         P_func(x_eval),
         's',
         markersize=8,
         label='x = 2.25')

plt.title("Interpolación de Lagrange",
          fontsize=15)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True)
plt.legend()

plt.show()

# =========================================================
# GRÁFICA 3
# SPLINES CÚBICOS
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(x, fx, 'o',
         markersize=8,
         label='Datos tabulados')

plt.plot(x_suave, y_spline,
         linewidth=2,
         label='Spline cúbico')

plt.plot(x_eval,
         spline(x_eval),
         's',
         markersize=8,
         label='x = 2.25')

plt.title("Método de Splines Cúbicos",
          fontsize=15)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True)
plt.legend()

plt.show()
```
