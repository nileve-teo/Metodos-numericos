<li>Regla de Simpson 1/3 Compuesta (n = 12).</li>
<li>Regla de Simpson 3/8 Compuesta (n = 12).</li>
<li>Cuadratura de Gauss-Legendre con m = 3 puntos.</li>

```python
import numpy as np
from scipy.interpolate import CubicSpline

# ==========================
# DATOS
# ==========================
V = np.array([1,1.25,1.5,1.75,2,2.25,2.5,2.75,3,3.25,3.5,3.75,4])
P = np.array([300.2,242.1,201.5,169.8,151,134.3,120.8,
              107.9,99.1,93.4,86.2,79.5,75.1])

# ==========================
# SPLINE (para Gauss)
# ==========================
spline = CubicSpline(V, P)

# ==========================
# TRAPECIO COMPUESTO
# ==========================
def trapecio(x, y):
    h = x[1] - x[0]
    return h * ( (y[0] + y[-1]) / 2 + np.sum(y[1:-1]) )

# ==========================
# SIMPSON 1/3
# ==========================
def simpson_13(x, y):
    h = x[1] - x[0]
    return (h/3) * (y[0] + y[-1] +
                    4*np.sum(y[1:-1:2]) +
                    2*np.sum(y[2:-1:2]))

# ==========================
# SIMPSON 3/8
# ==========================
def simpson_38(x, y):
    h = x[1] - x[0]
    s3, s2 = 0, 0

    for i in range(1, len(y)-1):
        if i % 3 == 0:
            s2 += y[i]
        else:
            s3 += y[i]

    return (3*h/8) * (y[0] + y[-1] + 3*s3 + 2*s2)

# ==========================
# GAUSS LEGENDRE (orden 3)
# ==========================
def gauss_legendre_3(f, a, b):

    xi = np.array([-np.sqrt(3/5), 0, np.sqrt(3/5)])
    wi = np.array([5/9, 8/9, 5/9])

    suma = 0
    for x, w in zip(xi, wi):
        t = ((b-a)/2)*x + (a+b)/2
        suma += w * f(t)

    return (b-a)/2 * suma

# ==========================
# RESULTADOS
# ==========================
W_trap = trapecio(V, P)
W_s13  = simpson_13(V, P)
W_s38  = simpson_38(V, P)
W_gauss = gauss_legendre_3(spline, V[0], V[-1])

# ==========================
# ERRORES (respecto a Gauss)
# ==========================
err_trap = abs((W_trap - W_gauss)/W_gauss)*100
err_s13  = abs((W_s13  - W_gauss)/W_gauss)*100
err_s38  = abs((W_s38  - W_gauss)/W_gauss)*100
err_gauss = 0

# ==========================
# TABLA TIPO SLIDE (FORMATO EXACTO)
# ==========================

print("\nRESULTADOS DE INTEGRACIÓN NUMÉRICA\n")

header = f"{'MÉTODO':<18} | {'VALOR APROXIMADO':<18} | {'ERROR RELATIVO %':<18}"
print(header)
print("-"*len(header))

print(f"{'Trapecio':<18} | {W_trap:<18.4f} | {err_trap:<18.4f}")
print(f"{'Simpson 1/3':<18} | {W_s13:<18.4f} | {err_s13:<18.4f}")
print(f"{'Simpson 3/8':<18} | {W_s38:<18.4f} | {err_s38:<18.4f}")
print(f"{'Gauss-Legendre':<18} | {W_gauss:<18.4f} | {err_gauss:<18.4f}")
```

<p style="text-align:justify;">
Para la Cuadratura de Gauss-Legendre, debido a que no se dispone de una función analítica continua sino de un conjunto discreto de datos experimentales, se utiliza una técnica de interpolación numérica mediante <b>spline cúbico</b>, implementado en Python con <b>scipy.interpolate.CubicSpline</b>.
</p>

<p style="text-align:justify;">
Este procedimiento permite reconstruir una función continua y suavemente derivable \( P(V) \), a partir de los 13 puntos experimentales dados. Una vez obtenida dicha función interpolada, se aplica el método de <b>Gauss-Legendre con m = 3 puntos</b> en el intervalo \([V_{min}, V_{max}]\).
</p>

<p style="text-align:justify;">
La integral evaluada corresponde a:
</p>

<center>
<font size="4" color="blue">
$$W = \int_{V_{min}}^{V_{max}} P(V)\, dV$$
</font>
</center>

<p style="text-align:justify;">
El uso de la interpolación permite transformar datos discretos en una representación continua, facilitando la aplicación de métodos de cuadratura de alta precisión como Gauss-Legendre.
</p>

```python
import numpy as np
import pandas as pd
from scipy.interpolate import CubicSpline

# ==========================
# DATOS
# ==========================

V = np.array([1,1.25,1.5,1.75,2,2.25,2.5,2.75,3,3.25,3.5,3.75,4])

P = np.array([300.2,242.1,201.5,169.8,151,134.3,120.8,
              107.9,99.1,93.4,86.2,79.5,75.1])

n = len(V)-1
h = (V[-1]-V[0])/n

# ==========================
# TRAPECIO COMPUESTO
# ==========================

def trapecio_compuesto(x,y):

    h = x[1]-x[0]

    return h*((y[0]+y[-1])/2 + np.sum(y[1:-1]))

# ==========================
# SIMPSON 1/3 COMPUESTO
# ==========================

def simpson_13(x,y):

    h = x[1]-x[0]

    suma_impares = np.sum(y[1:-1:2])
    suma_pares   = np.sum(y[2:-1:2])

    return (h/3)*(y[0] + y[-1]
                  + 4*suma_impares
                  + 2*suma_pares)

# ==========================
# SIMPSON 3/8 COMPUESTO
# ==========================

def simpson_38(x,y):

    h = x[1]-x[0]

    suma3 = 0
    suma2 = 0

    for i in range(1,len(y)-1):

        if i % 3 == 0:
            suma2 += y[i]
        else:
            suma3 += y[i]

    return (3*h/8)*(y[0] + y[-1]
                    + 3*suma3
                    + 2*suma2)

# ==========================
# SPLINE CÚBICO
# ==========================

spline = CubicSpline(V,P)

# ==========================
# GAUSS LEGENDRE m = 3
# ==========================

def gauss_legendre_3(f,a,b):

    xi = np.array([
        -np.sqrt(3/5),
         0,
         np.sqrt(3/5)
    ])

    wi = np.array([
        5/9,
        8/9,
        5/9
    ])

    suma = 0

    for x,w in zip(xi,wi):

        t = ((b-a)/2)*x + (a+b)/2

        suma += w*f(t)

    return ((b-a)/2)*suma

# ==========================
# RESULTADOS
# ==========================

W_trap = trapecio_compuesto(V,P)

W_s13  = simpson_13(V,P)

W_s38  = simpson_38(V,P)

W_gauss = gauss_legendre_3(spline,V[0],V[-1])

# ==========================
# ERRORES
# ==========================

err_trap = abs(W_trap-W_gauss)/abs(W_gauss)*100

err_s13 = abs(W_s13-W_gauss)/abs(W_gauss)*100

err_s38 = abs(W_s38-W_gauss)/abs(W_gauss)*100

print("\n")
print("\nGRÀFICAS\n")
import numpy as np
import matplotlib.pyplot as plt

# Valores para la curva suave
x_fino = np.linspace(V.min(), V.max(), 500)

plt.figure(figsize=(12,7))

# Fondo de la figura
plt.gcf().patch.set_facecolor('#BB7C3E')

# Fondo del área de graficación
ax = plt.gca()
ax.set_facecolor('#D59A63')

# Curva spline
plt.plot(
    x_fino,
    spline(x_fino),
    color='#4B2E19',
    linewidth=3,
    label='Spline Cúbico'
)

# Puntos experimentales
plt.scatter(
    V,
    P,
    s=120,
    color='#E9D2BB',
    edgecolors='black',
    linewidths=1.2,
    zorder=3,
    label='Datos experimentales'
)

# Etiquetas de cada punto
for x, y in zip(V, P):

    plt.annotate(
        f'({x:.2f}, {y:.1f})',
        (x, y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center',
        fontsize=8,
        color='black',
        bbox=dict(
            boxstyle='round,pad=0.2',
            fc='#F7E7CE',
            ec='#8B5A2B',
            alpha=0.85
        )
    )

plt.title(
    'Interpolación mediante Spline Cúbico',
    fontsize=20,
    fontweight='normal',
    color='#F4B675'
)

plt.xlabel(
    'Volumen V (L)',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Presión P (kPa)',
    fontsize=12,
    color='white'
)

plt.grid(
    True,
    linestyle='--',
    alpha=0.4
)

plt.legend()

plt.tight_layout()

plt.show()

# ==========================
# GRAFICO:


# ==========================
# GRAFICO: ÁREAS POR MÉTODO
# ==========================

plt.figure(figsize=(12,7))
plt.gcf().patch.set_facecolor('#BB7C3E')
ax = plt.gca()
ax.set_facecolor('#D59A63')

x_fino = np.linspace(V.min(), V.max(), 500)
y_fino = spline(x_fino)

# ==========================
# CURVA BASE
# ==========================
plt.plot(x_fino, y_fino, color='#4B2E19', linewidth=3, label='Spline Cúbico')

# ==========================
# TRAPECIO (aproximación visual)
# ==========================
plt.fill_between(V, P, color='#F7E7CE', alpha=0.6, label='Trapecio (aprox)')

# ==========================
# SIMPSON 1/3 (resaltado por tramos pares/impares)
# ==========================
plt.plot(V, P, color='blue', linestyle='--', alpha=0.6, label='Simpson 1/3 (base visual)')

# sombreado parcial para Simpson 1/3
for i in range(0, len(V)-2, 2):
    xs = V[i:i+3]
    ys = P[i:i+3]
    plt.fill_between(xs, ys, alpha=0.3, color='blue')

# ==========================
# SIMPSON 3/8 (bloques de 3 intervalos)
# ==========================
plt.plot(V, P, color='green', linestyle=':', alpha=0.6, label='Simpson 3/8 (base visual)')

for i in range(0, len(V)-3, 3):
    xs = V[i:i+4]
    ys = P[i:i+4]
    plt.fill_between(xs, ys, alpha=0.25, color='green')

# ==========================
# DATOS EXPERIMENTALES
# ==========================
plt.scatter(V, P, color='white', edgecolors='black', s=80, zorder=5)

# ==========================
# ESTILO
# ==========================
plt.title("Comparación visual de métodos de integración", color='#F4B675', fontsize=16)
plt.xlabel("Volumen V (L)")
plt.ylabel("Presión P (kPa)")
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()


# ==========================
# GRAFICO EXTRA
# ==========================
plt.figure(figsize=(12,7))

plt.gcf().patch.set_facecolor('#BB7C3E')
ax = plt.gca()
ax.set_facecolor('#D59A63')

# Curva spline
plt.plot(
    x_fino,
    spline(x_fino),
    color='#4B2E19',
    linewidth=3,
    label='Spline Cúbico'
)

# Área bajo la curva
plt.fill_between(
    x_fino,
    spline(x_fino),
    color='#E9D2BB',
    alpha=0.6,
    label=f'Área = Trabajo ≈ {W_gauss:.2f} kPa·L'
)

# Datos originales
plt.scatter(
    V,
    P,
    s=120,
    color='white',
    edgecolors='black',
    zorder=3,
    label='Datos experimentales'
)

# Mostrar algunos valores
for x,y in zip(V,P):
    plt.annotate(
        f'{y:.1f}',
        (x,y),
        textcoords="offset points",
        xytext=(0,8),
        ha='center',
        fontsize=8
    )

# Texto dentro de la gráfica
plt.text(
    2.8,
    max(P)*0.75,
    f'W ≈ {W_gauss:.2f} kPa·L',
    fontsize=12,
    bbox=dict(
        facecolor='#F7E7CE',
        edgecolor='black'
    )
)

plt.title(
    'Trabajo Termodinámico: Área Bajo la Curva P(V)',
    fontsize=18,
    color='#F4B675',
    fontweight='bold'
)

plt.xlabel('Volumen V (L)', fontsize=12, color='white')
plt.ylabel('Presión P (kPa)', fontsize=12, color='white')

plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()

# ==========================
# GRAFICO 2
# ==========================

plt.figure(figsize=(12,7))

plt.gcf().patch.set_facecolor('#BB7C3E')

ax = plt.gca()
ax.set_facecolor('#D59A63')

barras = plt.bar(
    metodos,
    resultados,
    color='#E9D2BB',
    edgecolor='black',
    linewidth=1.5
)

for barra in barras:

    altura = barra.get_height()

    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{altura:.2f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold',
        color='black'
    )

plt.title(
    'Comparación de Métodos de Integración',
    fontsize=18,
    fontweight='normal',
    color='#F4B675'
)

plt.xlabel(
    'Método Numérico',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Trabajo W (J)',
    fontsize=12,
    color='white'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()

plt.show()


# ==========================
# GRAFICO 3
# ==========================

errores = [
    err_trap,
    err_s13,
    err_s38,
    0
]

plt.figure(figsize=(12,7))

plt.gcf().patch.set_facecolor('#BB7C3E')

ax = plt.gca()
ax.set_facecolor('#D59A63')

barras = plt.bar(
    metodos,
    errores,
    color='#E9D2BB',
    edgecolor='black',
    linewidth=1.5
)

for barra in barras:

    altura = barra.get_height()

    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{altura:.4f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold',
        color='black'
    )

plt.title(
    'Comparación de Errores Relativos',
    fontsize=18,
    fontweight='normal',
    color='#F4B675'
)

plt.xlabel(
    'Método Numérico',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Error Relativo (%)',
    fontsize=12,
    color='white'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()

plt.show()
```

<h3 style="color:#F4B675;">

1. Influencia del ruido experimental
</h3>
<justify>
Los datos de presión empleados en este ejercicio fueron proporcionados por el enunciado y corresponden a un conjunto de mediciones simuladas que incluyen un leve ruido aleatorio asociado a vibraciones mecánicas del sistema. El objetivo es analizar cómo dicho ruido puede afectar el desempeño de los diferentes métodos de integración numérica utilizados para calcular el trabajo realizado por el pistón.
La Regla del Trapecio aproxima la función mediante segmentos lineales entre puntos consecutivos. Debido a esta aproximación simple, presenta una menor sensibilidad a pequeñas fluctuaciones presentes en los datos, aunque generalmente ofrece una precisión inferior cuando la función posee una curvatura significativa.
Por otro lado, los métodos de Simpson 1/3 y Simpson 3/8 utilizan aproximaciones polinómicas de orden superior. Estas aproximaciones permiten representar mejor la curvatura de la función y, bajo condiciones ideales, proporcionan una mayor precisión. Sin embargo, cuando los datos contienen ruido, las aproximaciones polinómicas pueden verse más afectadas por las variaciones locales, aumentando la sensibilidad del método frente a pequeñas perturbaciones.
Al comparar los resultados obtenidos, se observa que todos los métodos producen valores muy próximos para la integral. No obstante, las diferencias encontradas reflejan la distinta forma en que cada técnica aproxima el área bajo la curva presión-volumen.
Los datos de presión utilizados en este estudio contienen un leve ruido aleatorio debido a las vibraciones mecánicas presentes durante el proceso de adquisición. Dichas perturbaciones afectan de manera diferente a cada método numérico empleado para calcular el trabajo termodinámico.

</justify>

<h3 style="color:#F4B675;">
2. Interpolación mediante Spline Cúbico y Gauss-Legendre
</h3>
<justify>
La cuadratura de Gauss-Legendre requiere una función continua, condición que no se cumple directamente con datos experimentales discretos. Por ello, se empleó interpolación mediante spline cúbico, generando una función continua P(V) a partir de los datos medidos.

La principal ventaja de esta estrategia radica en que permite obtener una función suave que reproduce adecuadamente la tendencia general de los datos experimentales. Gracias a ello, la cuadratura de Gauss-Legendre puede aprovechar su elevada precisión teórica para estimar el área bajo la curva presión-volumen.
No obstante, la interpolación también introduce una fuente adicional de error. La función spline obtenida constituye una aproximación matemática de la relación real entre presión y volumen, por lo que cualquier desviación producida por el ruido experimental puede reflejarse en la curva interpolada y afectar el valor final de la integral.

</justify>
