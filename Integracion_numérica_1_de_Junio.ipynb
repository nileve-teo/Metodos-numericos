```python
import numpy as np
import pandas as pd
from scipy.interpolate import CubicSpline

# ==========================
# DATOS
# ==========================

V = np.array([1,1.25,1.5,1.75,2,2.25,2.5,2.75,3,3.25,3.5,3.75,4])

P = np.array([300.2,242.1,201.5,169.8,151,134.3,120.8,
              107.9,99.1,93.4,86.2,79.5,75.1])

n = len(V)-1
h = (V[-1]-V[0])/n

# ==========================
# TRAPECIO COMPUESTO
# ==========================

def trapecio_compuesto(x,y):

    h = x[1]-x[0]

    return h*((y[0]+y[-1])/2 + np.sum(y[1:-1]))

# ==========================
# SIMPSON 1/3 COMPUESTO
# ==========================

def simpson_13(x,y):

    h = x[1]-x[0]

    suma_impares = np.sum(y[1:-1:2])
    suma_pares   = np.sum(y[2:-1:2])

    return (h/3)*(y[0] + y[-1]
                  + 4*suma_impares
                  + 2*suma_pares)

# ==========================
# SIMPSON 3/8 COMPUESTO
# ==========================

def simpson_38(x,y):

    h = x[1]-x[0]

    suma3 = 0
    suma2 = 0

    for i in range(1,len(y)-1):

        if i % 3 == 0:
            suma2 += y[i]
        else:
            suma3 += y[i]

    return (3*h/8)*(y[0] + y[-1]
                    + 3*suma3
                    + 2*suma2)

# ==========================
# SPLINE CÚBICO
# ==========================

spline = CubicSpline(V,P)

# ==========================
# GAUSS LEGENDRE m = 3
# ==========================

def gauss_legendre_3(f,a,b):

    xi = np.array([
        -np.sqrt(3/5),
         0,
         np.sqrt(3/5)
    ])

    wi = np.array([
        5/9,
        8/9,
        5/9
    ])

    suma = 0

    for x,w in zip(xi,wi):

        t = ((b-a)/2)*x + (a+b)/2

        suma += w*f(t)

    return ((b-a)/2)*suma

# ==========================
# RESULTADOS
# ==========================

W_trap = trapecio_compuesto(V,P)

W_s13  = simpson_13(V,P)

W_s38  = simpson_38(V,P)

W_gauss = gauss_legendre_3(spline,V[0],V[-1])

# ==========================
# ERRORES
# ==========================

err_trap = abs(W_trap-W_gauss)/abs(W_gauss)*100

err_s13 = abs(W_s13-W_gauss)/abs(W_gauss)*100

err_s38 = abs(W_s38-W_gauss)/abs(W_gauss)*100

# ==========================
# TABLA FORMAL
# ==========================

tabla = pd.DataFrame({
    "Método":[
        "Trapecio Compuesto",
        "Simpson 1/3",
        "Simpson 3/8",
        "Gauss-Legendre"
    ],
    "Trabajo [J]":[
        W_trap,
        W_s13,
        W_s38,
        W_gauss
    ],
    "Error Relativo [%]":[
        err_trap,
        err_s13,
        err_s38,
        0
    ]
})

tabla["Trabajo [J]"] = tabla["Trabajo [J]"].round(4)
tabla["Error Relativo [%]"] = tabla["Error Relativo [%]"].round(4)

print("\nRESULTADOS DE LA INTEGRACIÓN\n")


display(tabla)

print("\n")
print("\nGRÀFICAS\n")
import numpy as np
import matplotlib.pyplot as plt

# Valores para la curva suave
x_fino = np.linspace(V.min(), V.max(), 500)

plt.figure(figsize=(12,7))

# Fondo de la figura
plt.gcf().patch.set_facecolor('#BB7C3E')

# Fondo del área de graficación
ax = plt.gca()
ax.set_facecolor('#D59A63')

# Curva spline
plt.plot(
    x_fino,
    spline(x_fino),
    color='#4B2E19',
    linewidth=3,
    label='Spline Cúbico'
)

# Puntos experimentales
plt.scatter(
    V,
    P,
    s=120,
    color='#E9D2BB',
    edgecolors='black',
    linewidths=1.2,
    zorder=3,
    label='Datos experimentales'
)

# Etiquetas de cada punto
for x, y in zip(V, P):

    plt.annotate(
        f'({x:.2f}, {y:.1f})',
        (x, y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center',
        fontsize=8,
        color='black',
        bbox=dict(
            boxstyle='round,pad=0.2',
            fc='#F7E7CE',
            ec='#8B5A2B',
            alpha=0.85
        )
    )

plt.title(
    'Interpolación mediante Spline Cúbico',
    fontsize=20,
    fontweight='normal',
    color='#F4B675'
)

plt.xlabel(
    'Volumen V (L)',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Presión P (kPa)',
    fontsize=12,
    color='white'
)

plt.grid(
    True,
    linestyle='--',
    alpha=0.4
)

plt.legend()

plt.tight_layout()

plt.show()

# ==========================
# GRAFICO EXTRA
# ==========================
plt.figure(figsize=(12,7))

plt.gcf().patch.set_facecolor('#BB7C3E')
ax = plt.gca()
ax.set_facecolor('#D59A63')

# Curva spline
plt.plot(
    x_fino,
    spline(x_fino),
    color='#4B2E19',
    linewidth=3,
    label='Spline Cúbico'
)

# Área bajo la curva
plt.fill_between(
    x_fino,
    spline(x_fino),
    color='#E9D2BB',
    alpha=0.6,
    label=f'Área = Trabajo ≈ {W_gauss:.2f} kPa·L'
)

# Datos originales
plt.scatter(
    V,
    P,
    s=120,
    color='white',
    edgecolors='black',
    zorder=3,
    label='Datos experimentales'
)

# Mostrar algunos valores
for x,y in zip(V,P):
    plt.annotate(
        f'{y:.1f}',
        (x,y),
        textcoords="offset points",
        xytext=(0,8),
        ha='center',
        fontsize=8
    )

# Texto dentro de la gráfica
plt.text(
    2.8,
    max(P)*0.75,
    f'W ≈ {W_gauss:.2f} kPa·L',
    fontsize=12,
    bbox=dict(
        facecolor='#F7E7CE',
        edgecolor='black'
    )
)

plt.title(
    'Trabajo Termodinámico: Área Bajo la Curva P(V)',
    fontsize=18,
    color='#F4B675',
    fontweight='bold'
)

plt.xlabel('Volumen V (L)', fontsize=12, color='white')
plt.ylabel('Presión P (kPa)', fontsize=12, color='white')

plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()

# ==========================
# GRAFICO 2
# ==========================

plt.figure(figsize=(12,7))

plt.gcf().patch.set_facecolor('#BB7C3E')

ax = plt.gca()
ax.set_facecolor('#D59A63')

barras = plt.bar(
    metodos,
    resultados,
    color='#E9D2BB',
    edgecolor='black',
    linewidth=1.5
)

for barra in barras:

    altura = barra.get_height()

    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{altura:.2f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold',
        color='black'
    )

plt.title(
    'Comparación de Métodos de Integración',
    fontsize=18,
    fontweight='normal',
    color='#F4B675'
)

plt.xlabel(
    'Método Numérico',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Trabajo W (J)',
    fontsize=12,
    color='white'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()

plt.show()


# ==========================
# GRAFICO 3
# ==========================

errores = [
    err_trap,
    err_s13,
    err_s38,
    0
]

plt.figure(figsize=(12,7))

plt.gcf().patch.set_facecolor('#BB7C3E')

ax = plt.gca()
ax.set_facecolor('#D59A63')

barras = plt.bar(
    metodos,
    errores,
    color='#E9D2BB',
    edgecolor='black',
    linewidth=1.5
)

for barra in barras:

    altura = barra.get_height()

    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{altura:.4f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold',
        color='black'
    )

plt.title(
    'Comparación de Errores Relativos',
    fontsize=18,
    fontweight='normal',
    color='#F4B675'
)

plt.xlabel(
    'Método Numérico',
    fontsize=12,
    color='white'
)

plt.ylabel(
    'Error Relativo (%)',
    fontsize=12,
    color='white'
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.4
)

plt.tight_layout()

plt.show()
```

<h3 style="color:#F4B675;">

1. Influencia del ruido experimental
</h3>

Los datos de presión utilizados en este estudio contienen un leve ruido aleatorio debido a las vibraciones mecánicas presentes durante el proceso de adquisición. Dichas perturbaciones afectan de manera diferente a cada método numérico empleado para calcular el trabajo termodinámico.

<br>

La Regla del Trapecio aproxima la función mediante segmentos lineales entre puntos consecutivos. Debido a esta aproximación sencilla, presenta una menor sensibilidad a pequeñas fluctuaciones locales de los datos, aunque generalmente proporciona una precisión inferior cuando la función presenta curvatura significativa.

<br><br>

Por el contrario, los métodos de Simpson 1/3 y Simpson 3/8 utilizan aproximaciones polinómicas de orden superior para representar la variación de la presión entre los puntos experimentales. Esta característica les permite alcanzar una mayor precisión cuando los datos son suaves y libres de perturbaciones. Sin embargo, en presencia de ruido experimental, los polinomios pueden ajustarse a variaciones locales que no representan el comportamiento físico real del sistema.

<br><br>

<h3 style="color:#F4B675;">
2. Interpolación mediante Spline Cúbico y Gauss-Legendre
</h3>

La cuadratura de Gauss-Legendre requiere una función continua para realizar la integración. Debido a que únicamente se disponía de un conjunto discreto de datos experimentales, fue necesario construir una representación continua de la presión en función del volumen utilizando interpolación mediante Spline Cúbico.

<br><br>

La principal ventaja de esta estrategia radica en que permite obtener una función suave que reproduce adecuadamente la tendencia general de los datos experimentales. Gracias a ello, la cuadratura de Gauss-Legendre puede aprovechar su elevada precisión teórica para estimar el área bajo la curva presión-volumen.

<br><br>

No obstante, la interpolación también introduce una fuente adicional de error. La función spline obtenida constituye una aproximación matemática de la relación real entre presión y volumen, por lo que cualquier desviación producida por el ruido experimental puede reflejarse en la curva interpolada y afectar el valor final de la integral.

**Impacto del Ruido en los Métodos de Orden Superior (Simpson) vs. Regla del Trapecio:**

Los métodos de integración de Simpson (1/3 y 3/8) son de orden superior que la Regla del Trapecio, lo que significa que, en teoría, ofrecen una mayor precisión para funciones suaves. Sin embargo, en presencia de ruido aleatorio en los datos experimentales, esta ventaja puede verse comprometida. Los métodos de Simpson ajustan polinomios de orden superior (cuadráticos para 1/3, cúbicos para 3/8) a los puntos de datos. Si los datos tienen ruido, estos polinomios pueden oscilar de forma no representativa entre los puntos, lo que puede amplificar el efecto del ruido y llevar a una menor precisión de lo esperado en comparación con el método del Trapecio, que es más robusto al ruido debido a su aproximación lineal.

En este caso, se observa que el error relativo de la regla del Trapecio (0.1104%) es menor que el de Simpson 1/3 (0.2735%) y Simpson 3/8 (0.2014%). Esto sugiere que el ruido presente en los datos experimentales afectó más a los métodos de Simpson, que son más sensibles a las fluctuaciones de los datos, mientras que el Trapecio, al ser más simple, fue menos susceptible a estas variaciones aleatorias.

**Ventajas y Desventajas de la Interpolación para Gauss-Legendre:**

La Cuadratura de Gauss-Legendre es un método muy preciso para integrar funciones continuas, especialmente polinomios. Sin embargo, su aplicación directa requiere una función analítica. Cuando solo se dispone de un conjunto discreto de puntos, como en este caso, se recurre a la interpolación para crear una función continua transitoria $P(V)$.

*   **Ventajas de la Interpolación (Spline Cúbico):**
    *   **Permite aplicar Gauss-Legendre:** Sin una función continua, la cuadratura de Gauss-Legendre no podría aplicarse. El Spline Cúbico crea una función suave y diferenciable que pasa a través de todos los puntos de datos, lo que es ideal para este método de integración de alta precisión.
    *   **Mayor Precisión Local:** Los Splines Cúbicos tienen la capacidad de adaptarse bien a la forma general de los datos, incluso si tienen cierta curvatura, lo que puede resultar en una aproximación más precisa de la función subyacente en comparación con interpolaciones de orden inferior.
    *   **Robustez a Ruido (moderado):** Aunque un Spline Cúbico es sensible al ruido, tiende a ser más estable que un polinomio de alto grado que interpole todos los puntos (interpolación de Lagrange), el cual puede generar oscilaciones erráticas.

*   **Desventajas de la Interpolación (Spline Cúbico):**
    *   **Introducción de Errores por Modelado:** La función interpolada es una aproximación de la función real $P(V)$. Si el modelo del Spline Cúbico no captura perfectamente la verdadera relación entre $P$ y $V$, se introduce un error de modelado que se propaga al cálculo de la integral.
    *   **Sensibilidad al Ruido:** Aunque mejor que un polinomio de alto grado, un Spline Cúbico sigue siendo sensible al ruido. El ruido en los datos de entrada puede hacer que el spline oscile localmente de una manera que no refleja el comportamiento físico real, lo que afecta la precisión de la integral.
    *   **Complejidad Computacional:** La creación del Spline Cúbico y su evaluación en los puntos de Gauss-Legendre añade una capa de complejidad computacional que no existiría si se tuviera una función analítica directa.

En resumen, la estrategia de interpolación con Spline Cúbico fue esencial para poder aplicar la cuadratura de Gauss-Legendre, que es un método de alta precisión. Sin embargo, la precisión global de la integración de Gauss-Legendre dependerá directamente de la calidad de la interpolación y de cuán bien el Spline Cúbico representa la función real P(V) en presencia de ruido. En este caso, el método de Gauss-Legendre, al basarse en una interpolación suave, se considera la referencia más precisa para comparar los demás métodos.
